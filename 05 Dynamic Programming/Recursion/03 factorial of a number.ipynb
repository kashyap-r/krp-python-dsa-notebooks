```python
# factorial of a number n 
# is n (n-1) (n-2) (n-3)... (n-(n-1)) 
# ex. fact(4) = 4 * 3 * 2 * 1 = 24

def fact(n):
    if n == 1 :
        return n
    else:
        return (n * fact(n-1))
    
num = 4 

if num < 0:
    print ("Invalid Input! Enter a +ve number.")
elif num == 0:
    print ("Factorial of 0 is 1")
else: 
    print ("Factorinal of the number", num, "=", fact(num)) 
```

### List 

Can contain duplicate items
Mutable: items can be modified, replaced, or removed
Ordered: maintains the order in which items are added
Index-based: items are accessed using their position (starting from 0)
Can store mixed data types (integers, strings, booleans, even other lists)


```python
# Creating the list 
# Using square brackets 

a = ['a', 'b', 'c']
b = [1, 2, 3]
print (a)
print(b)

# Using list() contructor
a1 = list((1, 2, 3, 4, 5))
print (a1)
a2 = list('kashyap')
print (a2)
a3 = list()
print (a3)

n1 = [2] * 5
n2 = ['a'] * 6
print (n1)
print (n2)
n3 = [9.99] * 5
print (n3)
```

```python
# Accessing list elements  
# insertion or deletion from the beginning of the list is costly, all the elements 
# from the beiginning of the list needs to be shifted
# 

l1 = [1, 2, 3, 'a', ' ', '%']
print (l1)

print (l1[5])
print (l1[1:4])
print (l1[-1])
print (l1[:-1])

print (l1[::-1])
print (l1[-3:])
```

```python
# List methods 
# append(): adds an element to the end of the list
# extend(): adds multiple elements to the end of the list 
# insert(): adds and element at a specific position 
# clear(): removes all items 

# update the elements of te list 
a = [1, 2, 3, 4]
print (a) 
a[1] = 11
print (a) 
```

```python
# Removing elements from the list 
# remove(): removes the first occurance of an element : what if there are mulitple 
# occurances and I want to remove all of them 
# pop(): removes the element at a specific index or the last element if no index is 
# specified
# del statement deltes an element at a specific index

a = [1, 2, 3, 4, 5]
print (a) 
a.remove(2)
print (a) 
val = a.pop(3)
print (a) 

del a[0]
print (a)
```

```python
# Iterating over the lists

# using lops 

basket = ['apple','banana','cherry', 'donki', 'elegant']

for item in basket:
    print (item)
```

```python
# Nested LIsts 

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print (matrix)
```

How Python Stores List Elements?
In Python, a list doesn’t store actual values directly. Instead, it stores references (pointers) to objects in memory. This means numbers, strings and booleans are separate objects in memory and the list just keeps their addresses.

That’s why modifying a mutable element (like another list or dictionary) can change the original object, while immutables remain unaffected.

## Python Dictionary 

Python dictionary is a data structure that stores the value in key: value pairs. Values in a dictionary can be of any data type and can be duplicated, whereas keys can't be repeated and must be immutable.

Keys are case sensitive which means same name but different cases of Key will be treated distinctly.
Keys must be immutable which means keys can be strings, numbers or tuples but not lists.
Duplicate keys are not allowed and any duplicate key will overwrite the previous value.
Internally uses hashing. Hence, operations like search, insert, delete can be performed in Constant Time.
From Python 3.7 Version onward, Python dictionary are Ordered.


```python
## Dictionaries
d1 = {'Name': 'Kash', 1: [1, 2, 3, 4]}
print (d1)
print (d1.items())
print (d1.keys())
print (d1.values())

# Accessing values in a dict
print (d1['Name'])

print (d1.get(1))

d2 = {x: x**2 for x in [1, 2, 3, 4, 5]}
print (d2)
```

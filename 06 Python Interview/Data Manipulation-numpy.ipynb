```python
# Numpy array indexing - to be able to access the array
import numpy as np

ls = [11, 12, 13, 14, 15, 1.11]
arr = np.array(ls)

print (arr)
print (arr[0])

# length of the array
print (arr.shape)
arrRows = arr.shape[0]

print (arrRows)
# if its a 2-D array then
# arrCol = arr.shape[1]
# print (arrCol)
```

```python
# Slicing and Dicing the array
ls = [11, 12, 13, 14, 15, 1.11]
arr = np.array(ls)

print (arr[1:4])
print (arr[:])
arr1 = arr
arr1[:] = -1
print (arr1)

ls = [11, 12, 13, 14, 15, 1.11]
arr = np.array(ls)
arr2 = arr.copy()
print (arr2)
arr2[4] = 55
print (arr2)
```

```python
arr = np.arange(90)
print (arr), print()
arr = arr.reshape(9, 10)
print (arr)

print (arr[1, 1])
print (arr[1][1])
print ()
print (arr[5, 6])
print (arr[5][6])
```

```python
# Indexing multi-dimensional arrays
arr = np.arange(90)
arr = arr.reshape(9, 10)

print (arr)
# accessing is by rows by columns

rowLen = arr.shape[0]
colLen = arr.shape[1]
print (rowLen, colLen)
# Print all the elements of the row
for i in range(rowLen):
    for j in range(colLen):
        print (i, j, arr[i, j])
```

```python
# reshape the above array into
arr = arr.reshape(9, 2, 5)
# An array containint 9 rows, 2 columns with each column containing 5 elements each
print (arr)
```

```python
# Array Transpose
import numpy as np

arr = np.arange(60).reshape(6, 10)
print (arr), print()
arr.transpose()

arr = np.arange(60).reshape(6, 5, 2)
print (arr), print()
arr.transpose()
```

```python
# Some Statistical operations on Arrays and Matrices

import numpy as np

arr = np.random.randn(4, 4)

print (arr)
print (arr.var())

print(np.where(arr<0, 0, arr))

# find the mean
arr.mean()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

xx, yy = np.meshgrid(x, y, sparse=True)
z = np.sin(xx**2 + yy**2) / (xx**2 + yy**2)
h = plt.contourf(x, y, z)
```

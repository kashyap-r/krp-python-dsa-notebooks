## Dropping a Column or a row of a DataFrame

* use the function df.drop(labels, axis=0/1, inplace=True/False)
* So, when axis = 0 (default) it will look for the row with the label and delete the row
* Set axis = 1, it will search column wise and delete the column
* inplace = True, you dont need to store the result / output of the operation in another dataframe

```python
# dropping a column from a dataframe
anotherDF.drop('Q', axis=1)
# does not delete the row on its own, you have to assign it to a resulting dataframe
print (anotherDF)
resDF = anotherDF.drop('Q', axis=1)
print (resDF)
```

```python
print (anotherDF)
# anotherDF = anotherDF.drop('Q', axis=1, inplace=True)
anotherDF.drop('Q', axis=1, inplace=True)
print (anotherDF)
```

```python
# concatenating dataframes: Merging dta frames
import pandas as pd

raw_data1 = {
    'id': [1, 2, 3, 4, 5, 6],
    'city': ['Bengaluru', 'Pune', 'Chennai', 'Mumbai', 'New Delhi', 'Indore'],
    'rank': ['1st', '2nd', '3rd', '4th', '5th', '6th'],
    'population': ['11M', '4M', '10M', '14M', '20M', '6M']
}

raw_data2= {
    'id': [7, 8, 9, 10],
    'city': ['Berlin', 'Munich', 'London', 'Amsterdam'],
    'rank': ['21st', '32nd', '23rd', '44th'],
    'population': ['10M', '14M', '20M', '6M']
}

raw_data3 = {
    'id': [1, 4, 8, 9],
    'city': ['Dublin', 'Stockholm', 'Helsinki', 'Oslo'],
    'rank': ['11th', '16th', '43rd', '114th']
}

# Note: The ids overlap above
df1 = pd.DataFrame(raw_data1, columns=['id', 'city', 'rank', 'population'])
df2 = pd.DataFrame(raw_data2, columns=['id', 'city', 'rank', 'population'])
df3 = pd.DataFrame(raw_data3, columns=['id', 'city', 'rank'])
```

```python
# default concatenation, joining along the rows i.e. i.e. the number of rows increase and by default axis = 0
newdf = pd.concat([df1, df2])
print (newdf), print()

newdf = pd.concat([df1, df2], ignore_index=True)
print(newdf)
```

```python
# the concatenation happens along the columns, and missing rows / data is filled by NaN
newdf = pd.concat([df1, df2], axis=1)
print(newdf), print()
```

```python
# merging dataframes - can specify the join conditions
print (df2), print()
print (df3), print()
#  merges based on the intersection i.e. the new dataframe will have only the rows that are in common between the two dataframes
mdf1 = pd.merge(df2, df3, on='id')
print(mdf1), print()
```

```python
mdf1 = pd.merge(df1, df2, on='id')
print(mdf1), print()
# the new dataframe will be empty since there is no overlap or intersections between df1 and df2
```

```python
mdf1 = pd.merge(df2, df3, on='id', how='outer', indicator= True)
print(mdf1), print()
```

#### Note: The "how" parameter determines the output dataset, it can assume values 'left', 'right', 'inner', or 'outer'; with 'inner' being the default

```python
# Can also merge based on more than one column
mdf1 = pd.merge(df2, df3, on=['id', 'city'], how='outer', indicator= True)
print(mdf1), print()
```

```python
mdf1 = pd.merge(df2, df3, on=['id', 'city'], how='outer', indicator= True, suffixes=('_left', '_right'))
print(mdf1), print()
```

```python
mdf1 = pd.merge(df2, df3, left_index=True, right_index=True, suffixes=('_left', '_right'))
print(mdf1), print()
```

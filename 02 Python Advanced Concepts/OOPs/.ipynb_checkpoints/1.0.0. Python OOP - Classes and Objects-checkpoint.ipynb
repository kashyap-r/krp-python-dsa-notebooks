### Python Class 

- Class is a collection of objects
- Contains the blueprints or the prototype from which the objects are created
- Contains attributes and methods
- Created by keyword class
- attributes are variables belonging to the class
- **attributes are always public** and can be accessed using the dot (.) operator
Class Definition Syntax:

class ClassName:
    # Statement-1 
    .
    .
    .
    # Statement - n 
 

```python
# create an empty class 
class Automobile:
    pass
```

### Python Objects 

- Object is an entity that has a state and behavior associated with it
- An Object consists of -
  * State - Represented by the attributes of an objects, it also reflects the properties of an object
  * Behavior - Represented by the methods of an object. It also reflects the response of an object to other objects
  * Identity - It gives an unique name to an object and enables one object to interact with other objects.
- Example: If "Dog" is a class 
  * Identity can be considered the name of the dog like "Milo"
  * State or Attributes can be considered as the breed, age, or color of the dog
  * The behavior can be cosidered as to whether the dog is sleeping, eating, playing, etc.


```python
# This will create an object named milo of the class Dog
class Dog:
    pass

milo = Dog()
```

### The Python __init__ method

- is similar to constructors in C++ and Java
- is run as soon as an object of a class is instantiated
- useful to perform any initialization


```python
# create a class and object with class and instance atteibutes 

class Dog:
    # class attribute
    attr1 = "mammal"

    # instance attribute
    def __init__(self, name):
        self.name = name

# Object instantiation
Milo = Dog("Milo")
Zeus = Dog("Zeus")

# Accessing class attributes 
print ("Milo is a {}".format(Milo.__class__.attr1))
print ("Zeus is a {}".format(Zeus.__class__.attr1))

# Accessing instance attributes 
print("My name is {}".format(Milo.name))
print("My name is {}".format(Zeus.name))
```


### Creating Classes and Objects with methods

```python
class Dog:
    # class attribute 
    attr1 = "mammal"

    # instance attribute 
    def __init__(self, name):
        self.name = name 

    def speak(self):
        print ("My Name is {}".format(self.name))

# Object instantiation
dog1 = Dog("Milo")
Zeus = Dog("Zeus")

dog1.speak()
Zeus.speak()
```

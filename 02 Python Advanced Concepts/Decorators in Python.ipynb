### Decorators 

References:
1. https://www.geeksforgeeks.org/decorators-in-python/?ref=lbp

- DEcorators are a very powerful tool since it allows to modify the behavior of a function or calss
- allows us to wrap another function in order to extend the behavior or the wrapped function, without permanently modifying it.
  
Pre-req:
- First Class Objects: In Python functions are first class objects which means that functions can be used or passed as arguements
- Properties of first class functions
    * A function is an instance of the Object type
    * You can store the function in a variable
    * You can pass the function as a parameter to another function
    * You can return the function from a function
    * You can store them in data structures such as hash tables, lists, ...


```python
# Example-1: Treating Functions as Objects 

def shout(text):
    return text.upper()

print (shout("Hello"))
yell = shout 
print (yell("hee haaw"))
```

In the above example, we have assigned the function shout to a variable. This will not call the function instead it takes the function object referenced by a shout and creates a second name pointing to it, yell.

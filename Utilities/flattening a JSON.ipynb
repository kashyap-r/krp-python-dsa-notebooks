## Flattening a JSON File

Flattening a JSON file means converting its nested structure (dictionaries inside dictionaries, lists inside dictionaries, etc.) into a tabular format (rows and columns) so it can be easily analyzed with tools like pandas.


```python
import pandas as pd
import json

# Example JSON
data = {
    "id": 1,
    "name": "Alice",
    "address": {
        "city": "Bengaluru",
        "zip": "560001"
    },
    "orders": [
        {"id": 101, "product": "Laptop"},
        {"id": 102, "product": "Phone"}
    ]
}

# Flatten JSON
df = pd.json_normalize(data)

print(df)
```

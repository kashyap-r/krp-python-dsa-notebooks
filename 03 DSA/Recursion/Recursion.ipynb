4! = 4 * 3 * 2 * 1 
3! = 3 * 2 * 1
2! = 2 * 1

```python
def factorial(n):
    print (n)
    if n == 1 or n == 0:
        return 1 
    return n * factorial(n-1)

# print (factorial(0))

print (factorial(4))
```

```python
import traceback

def recursive_function(n):
    if n == 0:
        return
    # Print the current call stack
    print("Call stack at n =", n)
    traceback.print_stack()
    # Recursive call
    recursive_function(n - 1)

recursive_function(3)
```

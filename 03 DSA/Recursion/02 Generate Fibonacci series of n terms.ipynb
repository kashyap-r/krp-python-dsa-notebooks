```python
#  Fibonacci sequence is the integer sequence of 0, 1, 1, 2, 3, 5, 8â€¦. 

def fibo(n):
    if n <= 1:
        return n
    else:
        return (fibo(n-1) + fibo(n-2))
    
n_terms = 10 

if n_terms <= 0:
    print ("Invalid input!")
else:
    print ("Fibonacci Series:")
for i in range(n_terms):
    print(fibo(i))
```

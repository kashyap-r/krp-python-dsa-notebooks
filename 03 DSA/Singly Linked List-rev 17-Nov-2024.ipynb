```python
class Node:
    def __init__(self, data):
        self.data = data 
        self.next = None 

# List traversal 
def displayList(head):
    current = head
    while current:
        print (str(current.data)+'->', end=' ')
        current = current.next
    print ("End")

def searchList(head, target):
    while head is not None:
        if head.data == target:
            return True 
        head = head.next 
    return False

# finding length in singly linked list 
def getLength(head):
    length = 0 
    current = head
    while current is not None:
        length += 1
        current = current.next 
    return length 

# Insertion in Singly linked list 
# a. insertion in the beginning 
def insertAtStart(head, data):
    newNode = Node(data)
    newNode.next = head 
    head = newNode 
    return head

# b. Insertion at the end of a singly linked list 
def insertAtEnd(head, data):
    newNode = Node(data) 
    if head is None:
        print ("List empty !")
        return None
    current = head 
    while current.next is not None:
        current = current.next
    current.next = newNode
    return head

# c. Insertion at a specific position of Singly linked list 
def insertAtPosition(head, position, data):
    if position < 1:
        print ("Invalid position")
        return head
    newNode = Node(data)
    # to insert at the head position 
    if position == 1:
        newNode.next = head 
        return newNode
    # Now,traverse the list to find the node before insertion point
    prev = head 
    count = 1
    while count < position -1 and prev is not None:
        prev = prev.next 
        count += 1
    # if position is greater that no. of nodes 
    if prev is None:
        print ("Invalid position")
        return head 
    # insert new node at the specified position
    newNode.next = prev.next 
    prev.next = newNode 
    return head

# deletion in singly linked list 
# a. deletion at the beginning of the list 
def deleteFirstNode(head):
    if not head:
        print ("List Empty! cannot delete!")
        return None
    temp = head 
    head = head.next 
    temp = None 
    return  head 

def deleteLastNode(head):
    if not head:
        print ("List empty! cannot delete!")
        return None 
    if not head.next:
        print (f"Deleted the only node '{head.value}'in the list")
        head = None 
        return None 
    # Now find the second-to-last node of the list 
    secondToLast = head
    while secondToLast.next.next is not None:
        secondToLast = secondToLast.next
    # remove the last node
    secondToLast.next = None 
    return head 

# c. deletion at a specific position of a singly linked list 
# Python function to delete a node at a specific position
def deleteAtPosition(head, position):
    # If the list is empty or the position is invalid
    if head is None or position < 1:
        return head
    # If the head needs to be deleted
    if position == 1:
        temp = head
        head = head.next
        temp = None
        return head
    # Traverse to the node before the position to be deleted
    current = head
    for i in range(1, position - 1):
        if current is not None:
            current = current.next
    # If the position is out of range
    if current is None or current.next is None:
        return head
    # Store the node to be deleted
    temp = current.next
    # Update the links to bypass the node to be deleted
    current.next = current.next.next
    # Delete the node
    temp = None
    return head

    
# driver code 
head = None 
print (f"Insert at Start:")
head = insertAtStart(head, 3)
head = insertAtStart(head, 2)
head = insertAtStart(head, 1)
head = insertAtEnd(head, 5)
head = insertAtEnd(head, 6)
displayList(head)
value = 4
position = 4
print (f"Insert an element {value} at Position {position}: ")
head = insertAtPosition(head, position, value) 
displayList(head)
value = 11
print (f"Insert element {value} at the beginning of the list")
head = insertAtStart(head, value)
displayList(head)
print (f"Delete First node if the list")
head = deleteFirstNode(head)
displayList(head)
print (f"Delete last node of the list")
head = deleteLastNode(head)
displayList(head)
print (f"Insert into position")
position = 4
value = 44
head = insertAtPosition(head, position, value)
displayList(head)
head = deleteAtPosition(head, 4)
print (f"Print delete at position")
displayList(head)
print ("Done~!")
```

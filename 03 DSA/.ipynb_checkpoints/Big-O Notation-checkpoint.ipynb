### O(n)

Below n items were printed out ... means O(n) is n

```python
def print_items(n):
    for i in range(n):
        print (i)

    for j in range(n):
        print (j)

print_items(10)
```

Here the loop executes n * n times print n * n values 
So, it is O(n^2)

```python
def print_items(n):
    for i in range(n):
        for j in range(n):
           print (j)

print_items(10)
```

### Drop non-dominants 

In the below example 
the nested for loop "a" executes n * n operations 
while the for loop "b" has n operations

So the time complexity can ebe expressed as O(n^2 + n) 

Since n is insignificant or non-dominant compared to n^2, we can ignore n in the above notation 
So, the time complexity can be re-written as O(n^2) 




```python
#  Simplifiying the code for Big O

def print_items(n):
    # a
    for i in range(n):
        for j in range(n):
           print (i, j)
    # b
    for k in range(n):
        print (k)

print_items(10)
```

```python
def print_items(a, b):
    for i in range (a):
        print(i)

    for j in range(b):
        print (j)
# here we cannot assign O(n) to the first loop and O(n) to the second 
# So, it will be O(a + b) 

# If it is nested for loops like below, then it will be O(a*b) operations 
def print_items(a, b):
    for i in range (a):
        for j in range(b):
            print (i, j)
```

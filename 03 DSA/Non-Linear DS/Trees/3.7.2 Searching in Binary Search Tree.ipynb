### Searching in Binary Search Tree (BST)

Given a BST, task is to search a node in this BST. Since this tree is already sorted we can use the Binary Search Algorithm 



Output: True
Explanation: 8 is present in the BST as right child of root

Input:  Root of the below BST    



Steps / Algoroithm:
1. Start at the root 
2. Compare the value to be search with the value of the root
    - If it is equal, search element found
    - if it is smaller, search in the left subtree 
    - If it is larger, then search in the right subtree 
3. Repeat the step till no more traversal is possible 
4. If at any point the key is found, returh True else False. 
 

```python
class Node:
    def __init__(self, value):
        self.value = value 
        self.left = None 
        self.right = None 

def searchBST(root, key):
    # base case 
    if root is None or root.value == key:
        return root
    # if key is greater than the root, navigate to the right subtree
    if root.value < key:
        return searchBST(root.right, key)
    else:
        return searchBST(root.right, key)

root = Node(50)
root.left = Node(30)
root.right = Node(70)
root.left.left = Node(20)
root.left.right = Node(40)
root.right.left = Node(60)
root.right.right = Node(80)

print ("Found" if searchBST(root, 19) else "Not FOund")
print ("Found" if searchBST(root, 80) else "Not Found!!")
```

Time COmplexity: O(h), where h is the hieght of the tree 
Auxiliary Space: O(h), because of the space needed to store the recursion stack

### Iterative Approach to Searching in a Binary Search tree (BST) 

https://www.geeksforgeeks.org/iterative-searching-binary-search-tree/

Appraoch:
Start from the root node
If the current node's data is equal to the key, then return True 
If node's value is less thatn the key, then traverse the right subtree
Else, set the current left to traverse in the left subtree 
If current beccomes null, then key is not present in BST, return False


```python
class Node:
    def __init__(self, x):
        self.data = x 
        self.left = None 
        self.right = None 

def search(root, x):
    curr = root 
    while curr is not None:
        if curr.data == x:
            return True 
        elif curr.data < x:
            curr = curr.right
        else: 
            curr = curr.left 
    return False

def printInOrder(root):
    if root:
        printInOrder(root.left)
        print(root.data, end=' ')
        printInOrder(root.right)



root = Node(20)
root.left = Node(8)
root.left.left = Node(4)
root.left.right = Node(12)
root.left.right.left = Node(10)
root.left.right.right = Node(14)
root.right = Node(22)

printInOrder(root)

x = 12
print ()
print(search(root, x)) 
```

```python
# Python program to search in 
# a BST.
class Node:
    def __init__(self, x):
        self.data = x
        self.left = None
        self.right = None

# Function to search in a bst.
def search(root, x):
    
    curr = root
    
    while curr is not None:
        
        # If curr node is x
        if curr.data == x:
            return True
            
        # Search in right subtree
        elif curr.data < x:
            curr = curr.right
            
        # Search in left subtree
        else:
            curr = curr.left
    
    # If x is not found.
    return False

if __name__ == "__main__":
    
    # Create a hard coded BST.
    #        20
    #       /  \
    #      8   22
    #     / \
    #   4   12
    #       /  \
    #     10   14
    root = Node(20)
    root.left = Node(8)
    root.left.left = Node(4)
    root.left.right = Node(12)
    root.left.right.left = Node(10)
    root.left.right.right = Node(14)
    root.right = Node(22)
    
    x = 12
    print(search(root, x))
```

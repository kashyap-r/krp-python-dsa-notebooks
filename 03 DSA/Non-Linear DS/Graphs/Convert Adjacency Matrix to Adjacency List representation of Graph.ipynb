Given a adjacency matrix representation of a Graph. The task is to convert the given Adjacency Matrix to Adjacency List representation.






# Python 3 program to convert Adjacency matrix
# representation to Adjacency List
 
from collections import defaultdict
# converts from adjacency matrix to adjacency list
def convert(a):
    adjList = defaultdict(list)
    for i in range(len(a)):
        for j in range(len(a[i])):
                       if a[i][j] != 0:
                           adjList[i].append(j)
    return adjList
 
# driver code
a =[[0, 0, 1], [0, 0, 1], [1, 1, 0]] # adjacency matrix
AdjList = convert(a)
print("Adjacency List:")
# print the adjacency list
for i in AdjList:
    print(i, end ="")
    for j in AdjList[i]:
        print(" -> {}".format(j), end ="")
    print()

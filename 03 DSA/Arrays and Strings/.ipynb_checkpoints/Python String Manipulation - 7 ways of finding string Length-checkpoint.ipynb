```python
# Method #1
# Using the built-in funciton 

str = "Kashyap Raj"
print (len(str))
```

```python
# Method #2 
# Using loop 

class String:
    def findLength(self, str):
        counter = 0 
        for i in str:
            counter += 1
        return counter

cl = String()
arr = "Kashyap Rajpurohit"
print ("The length of the string is: ",cl.findLength(arr))
```

```python
# Method #3
# Using while loop and slicing 

# returns the length of the string 
def findLen(str):
    counter = 0
    while str[counter:]:
        counter += 1 
    return counter

str = "Kashyap Rajpurohit"
print (findLen(str))
```

```python
# Method #4 
# Using string methods join and count

def findLen(str):
    if not str:
        return 0
    else:
        some_random_str = "py"
        return some_random_str.join(str).count(some_random_str) + 1

str = "Kashyap Rajpurohit"
print (findLen(str))
        
```

### Method #5 
#### Using reduce 
"""
Reduce method is used to iterate over the string and return a result of collection element provided to the reduce function. We will iterate over the string character by character and count 1 to result each time.
"""

```python
import functools 

def findLen(string):
    return functools.reduce(lambda x, y: x+1, string, 0)

str = "Kashyap Rajpurohit"
print (findLen(str))
```

### Methods#6 
Using sum() and list comprehension function. 

We use the list comprehensio for iterating over te string and sum function to sum today characeters in the string. 


```python
def findLen(string):
    return sum(1 for i in string)

str = "Kashyap Rajpurohit"
print (findLen(str))
```

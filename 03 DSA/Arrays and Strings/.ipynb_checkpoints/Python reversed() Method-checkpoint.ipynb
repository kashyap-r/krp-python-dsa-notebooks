### Python reversed() Method

https://www.geeksforgeeks.org/python-reversed-function/

```python
cars = ['nano', 'swift', 'bolero']
reversed_cars = list(reversed(cars))
print (reversed_cars)
```

```python
# 1. Python reversed() with Built-In Sequence Objects

# For tuple
seqTuple = ('g', 'e', 'e', 'k', 's')
print(list(reversed(seqTuple)))

# For range
seqRange = range(1, 5)
print(list(reversed(seqRange)))
```

```python
# 2. Python reversed() with for loop in Python
str = "Kashyap Raj"
revStr= "jaR payhsaK"
for ch in reversed(str): 
    print (ch,end="")
print()
for ch in reversed(revStr): 
    print (ch,end="")
```

```python
# 3. Python reversed() in Python with custom objects

class gdg:
    vowels = ['a', 'e', 'i', 'o', 'u']

    def __reversed__(self):
        return reversed(self.vowels)

if __name__ == "__main__":
    obj = gdg()
    print (list(reversed(obj)))    
```

```python
# 4. Python reversed() method with List
vowels = ['a', 'e', 'i', 'o', 'u']
print(list(reversed(vowels)))
```

```python
# 5. Python reversed() method with string
str = "Geeksforgeeks"
print(list(reversed(str)))
```

```python
def prime(num: int) -> bool:
    if num <= 0:
        return False
    if num in [2, 3]:
        return True
    
```

```python
# some more functions
def is_prime(n):
	if n in [2, 3]:
		return True
	if (n == 1) or (n % 2 == 0):
		return False
	r = 3
	while r * r <= n:
		if n % r == 0:
			return False
		r += 2
	return True

# print(is_prime(78), is_prime(79))
print(is_prime(11), is_prime(79))
```

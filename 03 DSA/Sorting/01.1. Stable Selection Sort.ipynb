### Stable Selection Sort

https://www.geeksforgeeks.org/stable-selection-sort/?ref=lbp



```python
def stableSelectionSort(arr, n):
    for i in range(n):
        min_idx = i 
        # find the minimum eleement  in unsorted array
        for j in range(i+1, n):
            if arr[min_idx] > arr[j]:
                min_idx = j 
        
        # Now move the minimum element at current i 
        key = arr[min_idx]
        while min_idx > i:
            arr[min_idx] = arr[min_idx -1]
            min_idx -= 1
        arr[i] = key

def printArray(arr):
    for i in range(len(arr)):
        print ("%d" %arr[i], end=' ')
    print()

    for ele in arr:
        print (ele, end= ' ')

# driver code 
arr = [4, 5, 3, 2, 4, 1]
stableSelectionSort(arr, len(arr))
printArray(arr)
```

Time Complexity: O(N2) // since two nested loops are used the time taken by the algorithm to complete all operation is quadratic.

Auxiliary Space: O(1)// since no extra array is used so the space taken by the algorithm is constant

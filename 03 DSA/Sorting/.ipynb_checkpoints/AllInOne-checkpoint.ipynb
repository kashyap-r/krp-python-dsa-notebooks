```python
def selectionSortAsc(arr):
    n = len(arr)
    for i in range(n-1):
        min_idx = i 
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def selectionSortDesc(arr):
    n = len(arr)
    for i in range(n-1):
        max_idx = i 
        for j in range(i+1, n):
            if arr[j] > arr[max_idx]:
                max_idx = j 
        arr[i], arr[max_idx]= arr[max_idx], arr[i]

def printArray(arr):
    for val in arr:
        print (val, end=" ")
    print()


if __name__ == "__main__":
    arr = [64, 25, 12, 22, 11]
    print ("Input array:", end = ' ')
    printArray(arr)
    selectionSortAsc(arr)
    print ("Output array in Ascending order:", end = ' ')
    printArray(arr)
    selectionSortDesc(arr)
    print ("Output array in Descending order:", end = ' ')
    printArray(arr)
    
```


Complexity Analysis:
a. there are two for loops or nested loops, so .. O(n * n) = O(n^2)
b. auxiliary space : O(1) 



## Recursive Selection Sort 

https://www.geeksforgeeks.org/recursive-selection-sort/?ref=lbp

The Selection Sort essentially maintains two parts 
1. The first part that is sorted
2. The second part that is yet to be sorted
 
The algorithm works repeatedly finding the minimum element (considering ascending order) from the unsorted part and putting it at the end of the sorted part

The idea of a recursive solution is to one by one increment the sorted part and recursively call for the remaining part 


```python
def minIndex(arr, i, j):
    if i == j:
        return i
    k = minIndex(arr, i+1, j)
    return (i if arr[i] < arr[k] else k)

def recursiveSelectionSort(arr, n, index=0):
    # return when the starting index and array size are same 
    if index == n:
        return -1 
    k = minIndex(arr, index, n-1)
    if k!= index:
        arr[k], arr[index] = arr[index], arr[k]
    recursiveSelectionSort(arr, n, index+1)

def printArray(arr):
    for val in arr:
        print (val, end=" ")
    print()


if __name__ == "__main__":
    arr = [64, 25, 12, 22, 11]
    print ("Input array:", end = ' ')
    printArray(arr)
    recursiveSelectionSort(arr, len(arr))
    print ("Output array in Ascending order:", end = ' ')
    printArray(arr)
    
```

```python
Time and Space Complexity Analysis of Selection Sort
Time Complexity: O(n^2)
Auxiliry Space: O(n)
```


## Bubble Sort Algorithm 

ref: https://www.geeksforgeeks.org/bubble-sort-algorithm/?ref=lbp



```python
def bubbleSort(arr):
    n = len(arr)
    # traverse thru all elements 
    for i in range(n):
        swapped = False 
        # last i elements are already in place 
        for j in range(0, n-i-1):
            # traverse the array from 0 to n-i-1 
            # swap if the element found is greater than the next element 
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True 
        if swapped == False:
            break 

if __name__ == "__main__":
    arr = [23, 12, 3, 45, 9, 1, 19]
    bubbleSort(arr)
    print ("Sorted array")
    for i in arr:
        print (i, end = ' ')       
```

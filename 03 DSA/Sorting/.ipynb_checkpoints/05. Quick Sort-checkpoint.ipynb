### QuickSort

Like Merge Sort, QuickSort is a Divide and Conquer algorithm. It picks an element as a pivot and partitions the given array around the picked pivot. There are many different versions of quickSort that pick pivot in different ways. 

    Always pick the first element as a pivot.
    Always pick the last element as a pivot (implemented below)
    Pick a random element as a pivot.
    Pick median as the pivot.
    
    
The key process in quickSort is a partition(). The target of partitions is, given an array and an element x of an array as the pivot, put x at its correct position in a sorted array and put all smaller elements (smaller than x) before x, and put all greater elements (greater than x) after x. All this should be done in linear time.



Partition Algorithm: 

There can be many ways to do partition, following pseudo-code adopts the method given in the CLRS book. The logic is simple, we start from the leftmost element and keep track of the index of smaller (or equal to) elements as i. While traversing, if we find a smaller element, we swap the current element with arr[i]. Otherwise, we ignore the current element. 

#### Pseudo Code for recursive QuickSort function:

/* low  –> Starting index,  high  –> Ending index */

quickSort(arr[], low, high) {

    if (low < high) {

        /* pi is partitioning index, arr[pi] is now at right place */

        pi = partition(arr, low, high);

        quickSort(arr, low, pi – 1);  // Before pi

        quickSort(arr, pi + 1, high); // After pi

    }

}

```python
# 24th Oct, 2024 
# Udemy Python DSA 

class QuickSort:

    def swap(self, arr, idx1, idx2):
        temp = arr[idx1]
        arr[idx1] = arr[idx2]
        arr[idx2] = temp
        
    def pivot(self, arr, pivot_idx, end_idx):
        # pick first item as the pivot 
        # copy all the elements < pivot value to the left side of the pivot and 
        # rest of them to the right 
        swap_idx = pivot_idx
        for i in range(pivot_idx + 1, end_idx + 1):
            if arr[i] < arr[pivot_idx]:
                swap_idx += 1 
                self.swap(arr, swap_idx, i)
        self.swap(arr, pivot_idx, swap_idx)
        return swap_idx
        
    def sort(self, arr, left, right):
        if left < right:
            pivot_idx = self.pivot(arr, left, right)
            self.sort(arr, left, pivot_idx - 1)
            self.sort(arr, pivot_idx + 1, right)
        return arr

    def quickSort(self, arr):
        return self.sort(arr, 0, len(arr)-1)

qs = QuickSort()
arr = [4, 6, 1, 7, 3, 2, 5]
print (qs.quickSort(arr))
```

**Linked List**

- Lnked list does not have indexes
- Linked list may not be spread across contiguous locations
- head - first node
- tail - last node
- one node points to the next

- Big-O for LInked Lists
    - Addition of a node at the end - O(1) 
    - Remove a node at the end of the list - O(n) 
    - Additin of a node at the beginning - O(1)
    - Deleting of a node at the beginnig - O(1)
    - Addition of a node in a random location of the list - O(n)
    - Deleting a node at a random location of the list - O(n)

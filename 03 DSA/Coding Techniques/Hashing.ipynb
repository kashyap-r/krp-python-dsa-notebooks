### Hashing in Data Structure 

- Hashing is atechnique used in daa structure to efficiently store and retrieve data for quick access
- involves mapping data to a specific index in a hash table using a hash function that enables fast retrieval of information based on key 
- commonly used in databases, caching systems, and to optimize search and retrieval operations
- can achieve (search, insert, delete) in O(1) time on average

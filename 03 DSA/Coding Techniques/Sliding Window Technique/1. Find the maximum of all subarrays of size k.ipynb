Given an array of integers, calculate the maximum sum of ‘k’ consecutive elements in the array.

Example 
Input  : arr[] = {100, 200, 300, 400}, k = 2
Output : 700


Considerations / Edge Cases / Constraints:

1. Input Array - whether there is a limitation of the size of the array
    - too large arrays may need LLs else a simple list may do the job
    - Whether the input array has only positive integers

2. Validation Criteria
    - Input Array
          * array of Zeros
          * Array of 1s
          * are decimals allowed
          * any possibility of character input 
    - window size: Is it fixed
          * What if window size is -ve or Zero
          * window size K >= N
3. Whether the max sum
    - Is sum of only +ve integers
    - Is the sum of any combination of sub-arrays including -ve numbers 
4. What is the output required ?
   - Is it just the Sum ?
   - Is we need to output the array that led to the answer ?

```python
# code
import sys

# O(n * k) solution for finding
# maximum sum of a subarray of size k
INT_MIN = -sys.maxsize - 1

# Returns maximum sum in a
# subarray of size k.


def maxSum(arr, n, k):

    # Initialize result
    max_sum = INT_MIN

    # Consider all blocks
    # starting with i.
    for i in range(n - k + 1):
        current_sum = 0
        for j in range(k):
            current_sum = current_sum + arr[i + j]

        # Update result if required.
        max_sum = max(current_sum, max_sum)

    return max_sum


# Driver code
arr = [1, 4, 2, 10, 2, 3, 1, 0, 20]
k = 4
n = len(arr)
print(maxSum(arr, n, k))
```

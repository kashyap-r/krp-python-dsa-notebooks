```python
# Prnt lisnked list with a cycle
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

def buildlinkedlist(values, pos=-1):
    """
    Build a linked list from a list of values.
    If pos >= 0, connect the tail node to the node at index `pos` to form a cycle.

    :param values: list of integers for node values
    :param pos: index to connect tail to (-1 = no cycle)
    :return: head of linked list
    """
    if not values:
        return None

    head = ListNode(values[0])
    nodes = [head]
    current = head

    for v in values [1:]:
        node = ListNode(v)
        current.next = node
        nodes.append(node)
        current = node

    if pos != -1:
        current.next = nodes[pos] #create cycle
    return head

def isPalindrome():
    if head is None or head.next is None:
        return True

    # find the middle using fast and slow pointers
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    prev = None
    curr = slow
    while curr:
        nextNode = curr.next
        curr.next = prev
        prev = curr
        curr = nextNode

    left, right = head, prev
    while right:
        if left.val != right.val:
            return False
        left = left.next
        right = right.next
    return True

def printList(head, max_nodes = 50):
    """
    Print the linked list values in order, stopping after `max_nodes` steps
    to avoid infinite loops in case of cycles.
    """

    current = head
    count = 0
    output = []
    while current and count < max_nodes:
        output.append(str(current.val))
        current = current.next
        count += 1
    if current:
        output.append("...[cycle?]")
    else:
        output.append("None")
    print(" -> ".join(output))

# 1) Linear list
head1 = buildlinkedlist([1, 2, 3, 4], pos=-1)
printList(head1)
# Output: 1 -> 2 -> 3 -> 4 -> None

# 2) List with cycle from last node back to index 1
head2 = buildlinkedlist([3, 2, 0, -4], pos=1)
printList(head2)
# Output: 3 -> 2 -> 0 -> -4 -> 2 -> 0 -> -4 -> 2 -> ...[cycle?]


```

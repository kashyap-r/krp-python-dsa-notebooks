```python
# Nested Loops

for i in range(1, 6):
    for j in range(i):
        print (i, end="")
    print ()
```

```python
# Loop control statements 

for ch in "geeksforgeeks":
    if ch == 'e' or ch == 's':
        continue 
    print ("Current letter:", ch)
```

```python
# Break Statement 

for ch in "geeksforgeeks":
    if ch == 'e' or ch == 's':
        break 
print ("Current letter:", ch)
```

```python
# passs statement - used to do nothing 
for ch in "geeksforgeeks":
    pass
print ("last character", ch)
```

```python
# here the code iterates over each item in the list and prints the item

fruits = ["apple", "orange", "kiwi"]
for fruit in fruits:

    print(fruit)
```

```python
# this code manuall iterates thru the list nd prints items using an iterator

fruits = ["apple", "banana", "orange", "kiwi"]
iterObj = iter(fruits)
while True:
    try:
        fruit = next (iterObj)
        print (fruit)
    except StopIteration:
        break   
```

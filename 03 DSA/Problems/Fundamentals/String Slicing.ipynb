```python
"""
String Slicing
- a wy to extract portion of sa string by specifying the start and end indexes
- syntax is string[start:end] where start is the starting index (a number) and end is the stopping index (excluded)
"""
```

```python
s = "IamWanderer"

print (s[1:4])
print (s[:5])
print (s[3:])
# reverses the string
print (s[::-1])
```

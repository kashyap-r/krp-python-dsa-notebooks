```python
# Split() and join()

"""
split()- splits the string into a list of strings after breaking the given string by specified operator

"""
```

```python
str = "One,Two,Three,Four,Five,Six"
words = str.split(',')
print (words)
words = str.split('o')
print (words)
```

```python
text = 'geeks for geeks'
# default splits at space
print(text.split())
print (text.split(' '))

text = 'geeks:for:geeks'
print (text.split(':'))

text = 'geeks:foer:geeks'
print (text.split('e'))

word = 'CatBatSatFatOr'
# Splitting at t
print(word.split('t'))
```

```python
"""
How does split work when maxsplit is specified?

The maxsplit parameter is used to control how many splits to return after the string is parsed. Even if there are multiple splits possible, it'll only do maximum that number of splits as defined by the maxsplit parameter.
"""
```

```python
word = 'geeks, for, geeks, Kashyap'
# maxsplit = 0
print (word.split(', ', 0))
# maxsplit = 4
print (word.split(', ', 4))
# maxsplit = 1
print (word.split(', ', 1))
```

```python
"""
parsing a string using split() methods

String parsing involves splitting a string into smaller segments based on a specific delimiter or pattern. This can be easily done by using a split() method in Python

"""
```

```python
text = "Hello geek, Welcome to Geeks forGeeks."

result = text.split()
print(result)
```

```python
"""
string COUNT() methods

Syntax:
string.count(substring, start = 0, end = len(s))

substring (required): The substring we want to count within the original string.
start (optional): The index position in the string where the search should begin. Default is 0.
end (optional): The index position in the string where the search should stop. Default is the length of the string (i.e., up to the end).

"""

s = "hello world"
res = s.count("o")
print(res)
```

```python
s = "Python is fun and Python is powerful."
print(s.count("Python"))

s = "GeeksforGeeks"
print(s.count("e"))

#Count Substring Occurrences with Start and End parameter
s = "apple banana apple grape apple"
substring = "apple"

# Using start and end parameters to count occurrences
# of "apple" within a specific range
res = s.count(substring, 1, 20)
print(res)
res = s.count(substring, 0, len(s)+1)
print(res)
```

```python
"""
Python String find() method

returns the index of the first occurance of a substring within a given string. If the substrig is not found, it returns -1

Syntax of find() method
s.find(substring, start, end))

Parameter:

substring: The substring to search for within the main string s.
start (optional): The starting index from where to begin the search.
end (optional): The ending index where the search should stop.
Return Value:

Returns the first index of the substring if it is found within the specified range.
Returns -1 if the substring is not found.

"""
```

```python
s = "abc   abc   abc"
res = s.find("abc")
print(res)
```

```python
"""
Python String join() method

The join() method in Python is used to concatenate the elements of an iterable (such as a list, tuple, or set) into a single string with a specified delimiter placed between each element.

Syntax of join()
separator.join(iterable)

Parameters:
separator: The string placed between elements of the iterable.
iterable: A sequence of strings (e.g., list, tuple etc) to join together.
Return Value:
Returns a single string formed by joining all elements in the iterable, separated by the specified separator
If the iterable contains any non-string values and it raises a TypeError exception.
"""
```

```python
# Joining a list of strings

a = "Joining a list of strings".split()
res = ' '.join(a)
res
```

```python
# Using join with tuples

s = ('Using', 'join', 'with', 'tuples')
res = "-".join(s)
print (s)
```

```python
# Using join() with set
s = {'Python', 'is', 'fun'}

res = "-".join(s)
print (res)
```

```python
# Using join() with Dictionary
"""
When using the join() method with a dictionary, it will only join the keys, not the values. This is because join() operates on iterables of strings and the default iteration over a dictionary returns its keys.
"""
d = {'Geek':1, 'for':2, 'Geeks':3}
res = '-'.join(d)
print (d)
print (res)
print ('-'.join(d.keys()))
# print ('-'.join(d.values()))
```

For example, 2 is written as II in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:


    I can be placed before V (5) and X (10) to make 4 and 9. 
    X can be placed before L (50) and C (100) to make 40 and 90. 
    C can be placed before D (500) and M (1000) to make 400 and 900.

Intuition:
In Roman NUmerals, when a  smaller value appears before a larger value, it represents 
subtraction, while when a smaller value appears after or equal to a larger value, it represents addition. 

Explanation:


Rules to Write Roman Numerals:

Order of Digits: Start with the largest values and work down (e.g., 2024 is MMXXIV, not MIM).

Subtractive Combinations: Use smaller digits before larger ones to indicate subtraction (e.g., IV for 4, IX for 9, XL for 40, XC for 90

Note: 
1. In subtractive combination, you cannot have more than one digit before the larger digit. For Ex. LXXX represents correct way of writing 80 and not XXC
2. So.. the number subtracting should be of the same order ... i.e. CD is valid to write 400 in roman, but XD is not valid to write 490 in roman numeral).

Repetition Limit: Do not repeat the same numeral more than three times in a row (e.g., use IV instead of IIII for 4).

```python
# from leetcode
class Solution:
    def romanToInt(self, s: str) -> int:
        m = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
        
        ans = 0        
        for i in range(len(s)):
            if i < len(s) - 1 and m[s[i]] < m[s[i+1]]:
                ans -= m[s[i]]
            else:
                ans += m[s[i]]
        
        return ans

r2i = Solution()
roman = 'III'
print (r2i.romanToInt(roman))
roman = 'XIXI'
print (r2i.romanToInt(roman))
```

```python
def roman_to_integer(roman):
    # Dictionary of Roman numeral values
    roman_values = {
        'I': 1, 'V': 5, 'X': 10, 'L': 50,
        'C': 100, 'D': 500, 'M': 1000
    }

    # Initialize the integer value and previous value
    integer_value = 0
    prev_value = 0

    for char in reversed(roman):
        current_value = roman_values[char]

        # If the current value is less than the previous value, subtract it
        if current_value < prev_value:
            integer_value -= current_value
        else:
            # Otherwise, add the current value
            integer_value += current_value

        # Update the previous value
        prev_value = current_value

    return integer_value

# Example usage:
print(roman_to_integer('MCMXCIV'))  # Outputs: 1994
print(roman_to_integer('IC')) 
```

```python
class Solution:
    def singleNumber(self, nums) -> int:
        dic = {}

        for i in range(len(nums)):
            if nums[i] not in dic:
                dic[nums[i]] = 1
            else:
                dic.pop(nums[i])

        return int(*dic)

s = Solution()
num = [4,1,2,1,2]

res = s.singleNumber(num)
print (res)
```

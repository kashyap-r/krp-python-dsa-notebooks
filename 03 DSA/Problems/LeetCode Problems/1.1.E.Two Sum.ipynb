```python
def twoSum(a, target):
    n = len (a) 
    for i in range(n):
        for j in range (i+1, n):
            # print (f"i={i}, j={j}, a[i]={a[i]}, a[j]={a[j]}")
            if a[i] + a[j] == target:
                return i, j

a = [3, 2, 4, 7, 8, 9]
print (f"Target reached with indices {twoSum(a, 15)}")
```

```python
# Brute force method
class Solution:
    def twoSum(self, nums, target: int):
        n = len(nums)        
        for i in range(n):
            for j in range (i+1, n):   
                # print (f"i={i}, j={j}")
                if nums[i] + nums[j] == target:
                    return [i, j]

sln = Solution()
a = [3, 2, 4, 7, 8, 9]
print (f"Target reached with indices {sln.twoSum(a, 15)}")
```

### Approch #2 Using a hash table 

1. Create an epty hash table to store elments and indices
2. Iterate through the array from left to right
3. For each element nums[i], calculate the complement by subtracting it from the target. complement = target - nums[i]
4. Check if the complement exist in the hash table. If it does, we have a solution
5. If the complement does not exist in the hash tabe, add the current element nums[i] to the hash table with its index as the value.
6. Repeat steps 3-5 until we find a solution or reach the end of the array.
7. If no solution is found return an empty or an appropriate indicator.

This approach has a time compleity of O(n) since the hash table lookups tke constant time on average. It allows us to solve this problem efficiently by making just one ass through the array.


```python
# Two pass has table 
class Solution:   
    def twoSum(self, nums, target):
        numMap = {}
        n = len(nums)
        # Build the hash table 
        for i in range(n):
            numMap[nums[i]] = i

        # find the complement 
        for i in range(n):
            complement = target - nums[i]
            if complement in numMap and numMap[complement] != i:
                return [i, numMap[complement]]

        return [] # no solution found 
            

if __name__ == "__main__":
    ts = Solution()
    a = [3, 2, 4, 7, 8, 9]
    target = 15 
    print (ts.twoSum(a, target))
```

```python
#  One-pass hash table
class Solution:
    def twoSum(self, nums, target):
        numMap = {}
        n = len(nums)

        for i in range(n):
            complement = target - nums[i]
            if complement in numMap:
                return [numMap[complement], i]
            numMap[nums[i]] = i
        return []  # No solution found

if __name__ == "__main__":
    ts = Solution()
    a = [3, 2, 4, 7, 8, 9]
    target = 15 
    print (ts.twoSum(a, target))
```

```python
class Solution:
    def twoSum(self, nums, target):
        pair_idx = {}

        for i, num in enumerate(nums):
            if target - num in pair_idx:
                return [i, pair_idx[target - num]]
            pair_idx[num] = i

if __name__ == "__main__":
    ts = Solution()
    a = [3, 2, 4, 7, 8, 9]
    target = 15 
    print (ts.twoSum(a, target))
```

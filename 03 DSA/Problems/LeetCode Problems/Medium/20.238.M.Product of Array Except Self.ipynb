Given an integer array nums, return an array such that answer[i] is equal to the product of all the elements of nums except nums[i]

Example 1:
arr = [1, 2, 3]
result = [6, 3, 2]

Example 2:
Input: nums = [1,2,3,4]
Output: [24,12,8,6

]

```python
# Self - rejected / failed  for a large amount of data 

def prodArray(nums):
    n = len(nums)
    prod = [1]*n
    loopCnt = 0
    while loopCnt < n:    
        for i in range(n):
            if i != loopCnt:
                prod[loopCnt] *= nums[i]                
        loopCnt += 1
    return prod

prodArray( [1, 2, 3, 4])
    
```

Prefix and suffix products are pretty neat concepts, often used in problems involving arrays. Let me break it down for you:

Prefix Products
A prefix product of an array is an array where each element at index i is the product of all elements in the original array from the start up to i. Heres a quick example:

For an array [1, 2, 3, 4], the prefix products would be:

At index 0: 1

At index 1: 1 * 2 = 2

At index 2: 1 * 2 * 3 = 6

At index 3: 1 * 2 * 3 * 4 = 24

So, the resulting prefix product array would be [1, 2, 6, 24].

Suffix Products
A suffix product of an array is an array where each element at index i is the product of all elements in the original array from i to the end. Using the same example array [1, 2, 3, 4], the suffix products would be:

At index 3: 4

At index 2: 3 * 4 = 12

At index 1: 2 * 3 * 4 = 24

At index 0: 1 * 2 * 3 * 4 = 24

So, the resulting suffix product array would be [24, 24, 12, 4].

Understanding these concepts can be super helpful in solving array-related problems, especially in competitive programming!

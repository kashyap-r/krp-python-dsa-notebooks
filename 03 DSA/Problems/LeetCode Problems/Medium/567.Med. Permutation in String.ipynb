## Extract all possible permutations of the same length of a string

### recursive approach 
Recursive Method: Generates all possible permutations by swapping characters, allowing us to directly manipulate the string.

#### Key Idea
The recursive method works by swapping characters in the string to generate different arrangements (permutations). At each step, one character is fixed, and the remaining characters are recursively permuted until all possible permutations are generated.

Steps of the Recursive Approach:
1. Base Case
   The recursion stops when the left index (l) equals the right index (r). This means we have fixed every character in the string, and the permutation is complete. At this point, the string is added to the result set.

:

```python

def permute(s, l, r, result):
    if l == r:
        result.add(s)  # Add the permutation to the set
    else:
        for i in range(l, r + 1):
            # Convert string to a list to perform swaps
            s_list = list(s)
            # Swap
            s_list[l], s_list[i] = s_list[i], s_list[l]  # Swap characters
            # Recur
            permute(''.join(s_list), l + 1, r, result)

def extract_permutations_recursive(word):
    result = set()  # Use a set to store unique permutations
    permute(word, 0, len(word) - 1, result)
    return result

# Example usage:
word = "stop"
result = extract_permutations_recursive(word)
print(result)
```

```python
def backtrack(s, path, used, result):
    if len(path) == len(s):
        result.add(''.join(path))  # Store the valid permutation
        return
    for i in range(len(s)):
        if used[i]:
            continue
        # Mark this character as used
        used[i] = True
        path.append(s[i])  # Add character to the current path
        backtrack(s, path, used, result)  # Recur
        path.pop()  # Backtrack
        used[i] = False  # Unmark character

def extract_permutations_backtracking(word):
    result = set()  # To store unique permutations
    used = [False] * len(word)  # Track used characters
    backtrack(word, [], used, result)
    return result

# Example usage:
word = "stop"
result = extract_permutations_backtracking(word)
print(result)
```

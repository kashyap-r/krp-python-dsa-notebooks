### 9.E.Palindrome Number

Description:

Given an integer x, return true if x is a palindrome, and false otherwise 

An integer is a palidrome when it reads the same forward and backward!!

Example 1: 
Input: x = 121
Output: true
Explanation: 121 reads as 121 from left to right and from right to left.

Example 2:

Input: x = -121
Output: false
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.

Example 3:

Input: x = 10
Output: false
Explanation: Reads 01 from right to left. Therefore it is not a palindrome.


Solution:

Two approaches to do it... 
Approach #1
    1. Convert to string 
    2. Reverse the string
    3. Compare the strings
    4. return result

Since it is a characeter by character comparison... accuracy will be more..

```python
### Approach #1
def palindrome(num):    
    numStr = str(num)
    revStr = numStr[::-1]
    if numStr == revStr:
        return True 
    else:
        return False

print (palindrome(11232114))
print (palindrome(-1123211))
print (palindrome(1123211.211))
print (palindrome(1123211))
# print (palindrome(01123211))
```

```python
# approach #2 
def palindrome(num):
    numDigits = len(str(num))
    div = num
    palin = 0
    for i in range(numDigits):        
        rem = div % 10
        quo = div // 10
        palin += rem * pow(10, numDigits-i-1)
        
        div = quo
    if num == palin:
        return True 
    else:
        return False
    

print (palindrome(1211))
print (palindrome(11211))
```

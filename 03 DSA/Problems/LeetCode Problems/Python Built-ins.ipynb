### enumerate()

- add a counter to each item in a List or other iterable 
- we can also turn it into a list of (number, item) pairs using list()


```python
a = ["Geeks", "for", "Geeks"]
b = enumerate(a)

for i, name in enumerate(a):
    print (f"Index {i}: {a}")

print (list(enumerate(a)))
print (b)
```

```python
arr = [1, 2, 3, 4, 5]
for index, value in enumerate(arr):
    print (f"{index}, {value}")
print ()
print (f"Starting the index from 1..")

# custom start index
for index, value in enumerate(arr, start=1):
    print (f"{index}, {value}")
  
```

```python

# Using Enumerate object in loops 
a = ['Kashyap', 'Raj', 'Rajpurohit']
print()
for ele in enumerate(a):
    print(ele)
```

```python
# Accessing the next element 
a = ['Kashyap', 'Raj', 'Purohit']
b = enumerate(a)

next_val = next(b)
print (next_val)
```

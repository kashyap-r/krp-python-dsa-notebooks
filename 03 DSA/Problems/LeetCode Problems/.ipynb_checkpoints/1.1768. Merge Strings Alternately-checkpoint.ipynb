```python
class Solution:
    def mergeAlternate(self, word1: str, word2: str) -> str:
        string = ""
        count = 0 
        while count < min(len(word1), len(word2)):
            string += word1[count]
            string += word2[count]
            count += 1

        return string + word1[count:] + word2[count:]

sol = Solution()
string1 = "Ksyp"
string2 = "aha"

string3 = ""

string3 = sol.mergeAlternate(string1, string2)
string3
    
```

```python
class solution:
    def mergeStr(self, str1, str2):
        result = ""
        maxlen = max(len(str1), len(str2))

        for i in range(maxlen):
            if (i < len(str1)):
                result += str1[i]
            if (i < len(str2)):
                result += str2[i]
        return result

sol = solution()
str1 = 'Ksyp'
str2 = 'aha'
str3 = ""
str3 = sol.mergeStr(str1, str2)
str3
```

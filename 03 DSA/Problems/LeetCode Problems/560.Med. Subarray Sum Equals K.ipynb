560. Subarray Sum Equals K

Given an array of integers nums and an integer k, return the total number of subarrays whose sum equals to k.

A subarray is a contiguous non-empty sequence of elements within an array.

Example 1:
Input: nums = [1,1,1], k = 2
Output: 2

Example 2:
Input: nums = [1,2,3], k = 3
Output: 2


```python
# get all combinations of subarrays from an array 
# using brute force method of two loops 

arr = [1, 2, 3]
n = len(arr)
subArr = set()

for i in range(n):
    for j in range(i, n):
        tempArr = tuple(arr[i:j+1])
        subArr.add(tempArr)

for row in subArr:
    print (list(row))
```

```python
# Using a sliding window and a set 

def getSubArrays(arr):
    uSubArr = set()
    n = len(arr)
    for start in range(n):
        tempArr = []
        for end in range(start, n):
            tempArr.append(arr[end])
            uSubArr.add(tuple(tempArr))

    for ele in uSubArr:
        print (list(ele))

arr = [1, 2, 3]
getSubArrays(arr)
```

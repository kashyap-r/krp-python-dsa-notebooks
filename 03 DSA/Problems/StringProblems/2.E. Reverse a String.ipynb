```python
"""
Reversing a String

Given a string s, the task is to reverse the string. Reversing a string means rearranging the characters such that the first character becomes the last, the second character becomes second last and so on.

Considerations
1. What of the string has special characters, numbers, spaces
2.




Input: s = "abdcfe"
Output: "efcdba"
Explanation: The first character a moves to last position, the second character b moves to second-last and so on.
"""
```

```python
"""
Method-1: Using backward traversal
Time Complexity: O(n)
Space Complexity: O(n)

- start from the last character of the string and move backward
- appending each character to a new string res.

"""
```

```python
# using a string
def reverseString(str):
    res = ""
    for i in range(len(str) - 1, -1, -1):
        res += str[i]
    return res

if __name__ == "__main__":
    str = "abdcfe"
    print(reverseString(str))

# using a list

def reverseString(str):
    res = []
    for i in range(len(str) - 1, -1, -1):
        res.append(str[i])
    return ''.join(res)

if __name__ == "__main__":
    str = "abdcfe"
    print(reverseString(str))
```

```python
# Method-2: Using two pointers
"""
O(n) time and O(1) space
"""

def reverseString(str):
    left = 0
    right = len(str) - 1
    # convert to a list before you proceed
    str = list(str)
    while left < right:
        str[left], str[right] = str[right], str[left]
        left += 1
        right -= 1

    return ''.join(str)

if __name__ == "__main__":
    str = "abdcfe"
    print(reverseString(str))
```

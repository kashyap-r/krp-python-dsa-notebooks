```python
"""
To check if string is a palindrome - Using String Slicing
"""
def isPalindrome(str):
    return str == str[::-1]
print(isPalindrome("1221"))
print(isPalindrome("Chamacha"))
print(isPalindrome("madam"))
print()
"""
To check if string is a palindrome - Using Two Pointer Method
(No Slicing)
"""
def isPalindrome(str):
    left, right = 0, len(str) - 1
    while left < right:
        if str[left] != str[right]:
            return False
        left += 1
        right -= 1
    return True

print(isPalindrome("1221"))
print(isPalindrome("Chamacha"))
print(isPalindrome("madam"))
print()
"""
Method - 3: Using Regular expressions
To ignore case and non-alphanumeric characters
"""
import re
def isPalindrome(str):
    s = re.sub(r"[^a-zA-Z0-9]", "", str).lower()
    return s == s[::-1]
print(isPalindrome("1221"))
print(isPalindrome("Chamacha"))
print(isPalindrome("madam"))
```

```python
"""
SAwap Two Numbers
Given Two Numbers a and b, the task is to swap them


1. Naive Approach - Using a third variable

2. Without using the third variable

3. Built-in swap
"""

```

```python
if __name__ == "__main__":
    a = 2
    b = 3
    print (f"a={a}, b={b}")
    temp = a
    a = b
    b = temp
    print (f"a={a}, b={b}")
    # here time complexity O(1), space" O(1)
    # method 2 - without using the third variable
    print (f"Second Method = without using third variable")
    a = 12
    b = 33
    print (f"a={a}, b={b}")
    a = a + b
    b = a - b
    a = a - b
    print (f"a={a}, b={b}")




```

```python
"""
Using Bitwise XOR
    a = a ^ b: Store the Bitwise XOR of a and b in a. Now, a holds the result of (a ^ b).
    b = a ^ b: Bitwise XOR the new value of a with b to get the original value of a. This gives us, b = (a ^ b) ^ b = a.
    a = a ^ b: Bitwise XOR the new value of a with the new value of b (which is the original a) to get the original value of b. This gives us, a = (a ^ b) ^ a = b.
"""

if __name__ == "__main__":
    a = 24
    b = 333
    print (f"Third Method = Using Bitwise XOR")
    print (f"a={a}, b={b}")
    a = a ^ b
    b = a ^ b
    a = a ^ b
    print (f"a={a}, b={b}")

"""
Time Complexity: O(1)
Auxiliary Space: O(1)
"""
```

```python
# 4th Method- Builtin swap
if __name__ == "__main__":
    a = 33
    b = 555
    print (f"a={a}, b={b}")
    a, b = b, a
    print (f"a={a}, b={b}")
```

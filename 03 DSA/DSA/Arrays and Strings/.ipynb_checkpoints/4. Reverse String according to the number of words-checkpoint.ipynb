### Reverse String according to the number of words

Problem Statement:
Given a string containing a number of words. 
If the count of words in string is even then reverse its even positionâ€™s words else reverse its odd position, push reversed words at the starting of a new string and append the remaining words as it is in order.

Examples: 
Input:  Ashish Yadav Abhishek Rajput Sunil Pundir
Output: ridnuP tupjaR vadaY Ashish Abhishek Sunil

Input:  Ashish Yadav Abhishek Rajput Sunil Pundir Prem
Output: merP linuS kehsihbA hsihsA Yadav Rajput Pundir 


To-Do 
1. Get the count of words in the sentence
2. if the count is even
3. then reverse the even position words and "push it into a stack"
4. else reverse the odd position words and "push it into a stack"
5. add the remaining words in the same order

Constraints:
1. limit on string length, is it to be scalable - then choice of data structure?
2. what happens to puncutation, special characters, multiple spaces
3. what if the input string is empty or a single word, or just two words ?
4. What if all words are identical ?
5. Does case matter?

1. Deque (Double-Ended Queue)
Using deque from the collections module can be more efficient for appending elements to either end. This is especially useful when constructing the result string in parts.

## perfect solution

```python
from collections import deque

def process_string_with_deque(s):
    words = s.split()
    n = len(words)
    result = deque()

    if n % 2 == 0:
        positions_to_reverse = range(1, n, 2)  # Even positions (1-based: 2, 4, ...)
    else:
        positions_to_reverse = range(0, n, 2)  # Odd positions (1-based: 1, 3, ...)

    for i in positions_to_reverse:
        result.appendleft(words[i][::-1])
    
    for i in range(n):
        if i not in positions_to_reverse:
            result.append(words[i])

    return ' '.join(result)

# Example usage:
s = 'Ashish Yadav Abhishek Rajput Sunil Pundir'
print(process_string_with_deque(s))  # Output: "owt one ruof three"
```

2. List
Python lists are a good choice because they provide efficient random access and are dynamic in size. You can easily iterate over words, reverse them, and construct the result.

## Solution not in order 

```python


def process_string(s):
    words = s.split()
    n = len(words)
    result = []

    if n % 2 == 0:
        positions_to_reverse = range(1, n, 2)  # Even positions (1-based: 2, 4, ...)
    else:
        positions_to_reverse = range(0, n, 2)  # Odd positions (1-based: 1, 3, ...)

    for i in positions_to_reverse:
        result.append(words[i][::-1])
    
    for i in range(n):
        if i not in positions_to_reverse:
            result.append(words[i])

    return ' '.join(result)

# Example usage:
s = "this is a test"
s = 'Ashish Yadav Abhishek Rajput Sunil Pundir'
print(process_string(s))  # Output: "si this tset a"
```

How to efficiently implement k stacks in a single array?

ref:
https://www.geeksforgeeks.org/efficiently-implement-k-stacks-single-array/?ref=lbp

We have discussed space-efficient implementation of 2 stacks in a single array.  

In this post, a general solution for k stacks is discussed. Following is the detailed problem statement. Create a data structure kStacks that represents k stacks. Implementation of kStacks should use only one array, i.e., k stacks should use the same array for storing elements. 


The following functions must be supported by k Stacks. push(int x, int sn) –> pushes x to stack number ‘sn’ where sn is from 0 to k-1 pop(int sn) –> pops an element from stack number ‘sn’ where sn is from 0 to k-1

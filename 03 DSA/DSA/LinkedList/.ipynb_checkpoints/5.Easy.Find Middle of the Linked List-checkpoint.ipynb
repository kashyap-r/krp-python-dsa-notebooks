Given a singly linked list, the task is to find the middle of the linked list. 

Analysis:
1. If the number of nodes are even, then there would be two middle nodes, so return the second middle node
    **Q: Ask the question if there is any specific criteria for this >?!**
2. Is the size of input taken as input ?
3. (Assumption) Whether it is a list of numbers

Input: linked list: 1->2->3->4->5
Output: 3 
Explanation: There are 5 nodes in the linked list and there is one middle node whose value is 3.


Input: linked list = 10 -> 20 -> 30 -> 40 -> 50 -> 60
Output: 40
Explanation: There are 6 nodes in the linked list, so we have two middle nodes: 30 and 40, but we will return the second middle node which 

is 40.
Edge Case Scenarios

Steps:
1. Create a linked list
2. get the count of the nodes of the list
3. find the mid-point as per the requirement :
1.  



[Naive Approach] By Counting Nodes – O(n) time and O(1) space:

The idea is to traversing the entire linked list once to count the total number of nodes. 
After determining the total count, traverse the list again and stop at the (count/2)th node to return its value. 

This method requires two passes through the linked list to find the middle element.

Time Complexity: O(2 * n) = O(n) where n is the number of nodes in the linked list.
Auxiliary Space: O(1)


```python
class node:
    def __init__(self, value):
        self.value = value
        self.next = None 

class LinkedList:
    def __def__(self):
        self.head = None 
        self.tail = None
```

```python
list = [10, 20, 30, 40, 50, 60]
mid_point = (7 // 2) 
print (mid_point)
list[mid_point]
```

```python
class Node:
    def __init__(self, x):
        self.data = x 
        self.next = None 

def getLength(head):
    length = 0 
    while head:
        length += 1 
        head = head.next

    return length 

def getMiddle(head):
    length = getLength(head)
    mid_index = length // 2
    while mid_index:
        head = head.next 
        mid_index -= 1 
    return head.data

def main():
    head = Node(10)
    head.next = Node(20)
    head.next.next = Node(30)
    head.next.next.next = Node(40)
    head.next.next.next.next = Node(50)
    head.next.next.next.next.next = Node(60)
    print (getMiddle(head))

if __name__ == "__main__":
    main()
```

#### [Expected Approach] 

## By Tortoise and Hare Algorithm – O(n) time and O(1) space:
##          OR Floyd's Cycle FInding Algorithm 



- Traverse the Linked list using a slow pointer (slow_ptr) and a fast pointer (fast_ptr).
- Move the slow pointer to the next node (one node forward) and the fast pointer to the next of the next nodes two nodes forward).
- When the fast pointer reaches the last node or NULL, then the slow pointer will reach the middle of the linked list.
- In caxe of the odd number of nodes in the list, the slow_ptr will reach the middle node when fast pointer reaches the last node..
- In case of even number of nodes, slow_ptr will reach the middle node when fast_ptr will become Null.   


```python
class Node:
    def __init__ (self, x):
        self.data = x
        self.next = None 

def getMiddle(head):
    slow_ptr = head
    fast_ptr = head

    while fast_ptr is not None and fast_ptr.next is not None:
        fast_ptr = fast_ptr.next.next
        slow_ptr = slow_ptr.next
    return slow_ptr.data 

def main():
    head = Node(10)
    head.next = Node(20)
    head.next.next = Node(30)
    head.next.next.next = Node(40)
    head.next.next.next.next = Node(50)
    head.next.next.next.next.next = Node(60)
    print (getMiddle(head))

if __name__ == "__main__":
    main()
```


### 1.Easy. Find the middle of the linked list 

Problem Description: Given a singly linked list, the task is to find the middle of the linked list. If the number of nodes are even, then there would be two middle nodes, so return the second middle node.

Example: 
Input: linked list: 1->2->3->4->5
Output: 3 
Explanation: There are 5 nodes in the linked list and there is one middle node whose value is 3.


Input: linked list = 10 -> 20 -> 30 -> 40 -> 50 -> 60
Output: 40
Explanation: There are 6 nodes in the linked list, so we have two middle nodes: 30 and 40, but we will return the second middle node which is 40.

Constraints:
1. 

**Approach #1: By Counting Nodes (Naive Approach)**

1. create the linked list 
2. add items
3. write a function to traverse the list and count the no. of elements
4. calculate the midpoint
5. traverse again to return the middle element

Time Complexity: O(n+n) = O(2n)
Auxiliary Space= O(1) 



```python
class Node:
    def __init__(self, x):
        self.data = x 
        self.next = None 

# helper function to find length of the linkked list
def getLength(head):
    length = 0 
    while head:
        length += 1 
        head = head.next 
    return length 

def getMidpoint(head):
    length = getLength(head)
    midPoint = length // 2 
    if midPoint % 2 == 0:
        midPoint = midPoint + 1 
    
    while midPoint:
        head = head.next 
        midPoint -= 1
    return head.data 
    
def main():
    head = Node(10)
    head.next = Node(20) 
    head.next
```

```python
# Calculating the mid point
import math
length = 7
midPoint = math.ceil((length + 1) / 2)
print(midPoint)
length = 8
midPoint = math.ceil((length + 1) / 2)
print(midPoint)
```

### **Singly Linked List Operations**
- Insertion
      * Insertion at the beginning - insertBegin()
      * Insertion at the end - insertEnd()
      * Insertion after a given node - insertAfter()
      * Insertion before a given node - insertBefore()
- Traversal
      * Linked list traversal 
      * Search for a given node based on value
      * Search for a given node based on index
  
- Deletion at
      * at the beginning of the list
      * after a given node
      * end of the linked list 
  

```python
### Approach #1 

class Node: 
    def __init__(self, data):
        self.data = data
        self.next = None 

def traverseList(head):
    if head is None:
        print ("List empty !")
        return -1 

    current = head 
    while current:
        print (str(current.data) + '->', end=" ")
        current = current.next
    print ("None")

def insertBegin(head, data):
    newNode = Node(data) 
    newNode.next = head 
    return newNode 

# driver Code 
head = None 
# Creates an instance 
head = insertBegin(head, 4) 
head = insertBegin(head, 3)
head = insertBegin(head, 2)
head = insertBegin(head, 1)

traverseList(head)
```

```python
#  Manual Calculation of word frequency 
class WordPro:
    def manualwordFreq(self, sentence):
        sentence = sentence.lower()
        words = sentence.split()

        freq_dict = {}
        for word in words:
            word = word.strip(",.!?:;")
            if word in freq_dict:
                freq_dict[word] += 1
            else:
                freq_dict[word] = 1
        return freq_dict

wp = WordPro()
sentence = "This is an xample sentence, this shows and this sentence shows an example of word frequencies."
frequencies = wp.manualwordFreq(sentence)
print (frequencies)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 

class WordFreq:
    def calWordFreq(self, string):
        vectorizer = TfidfVectorizer()
        vectors = vectorizer.fit_transform([string.lower()])

        feature_names = vectorizer.get_feature_names_out()
        freqs = vectors.toarray().flatten()
        return dict(zip(feature_names, freqs))

str = "This is an xample sentence, this shows and this sentence shows an xample of word frequencies."
wf = WordFreq()
frquencies = wf.calWordFreq(str)
print (frquencies)

        
    
```

```python
import spacy
from collections import Counter

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

def spacy_word_frequencies(sentence):
    # Process the sentence using spaCy
    doc = nlp(sentence.lower())
    
    # Get the text of each token and count frequencies
    word_freq = Counter(token.text for token in doc if token.is_alpha)
    
    return word_freq

# Example sentence
sentence = "This is an example sentence, and this sentence shows an example of word frequencies."

# Display word frequencies
frequencies = spacy_word_frequencies(sentence)
print(frequencies)
```

```python
import nltk
from nltk import FreqDist
import re

def nltk_word_frequencies(sentence):
    # Tokenize the sentence into words
    words = nltk.word_tokenize(sentence.lower())
    
    # Compute frequency distribution
    freq_dist = FreqDist(words)
    
    return freq_dist

# Example sentence
sentence = "This is an example sentence, and this sentence shows an example of word frequencies."

# Display word frequencies
frequencies = nltk_word_frequencies(sentence)
print(frequencies.most_common())
```

```python
import pandas as pd
import re

def pandas_word_frequencies(sentence):
    # Tokenize the sentence into words
    words = re.findall(r'\w+', sentence.lower())
    
    # Create a pandas DataFrame
    df = pd.DataFrame(words, columns=['word'])
    
    # Get word frequencies using `value_counts()`
    word_freq = df['word'].value_counts()
    
    return word_freq

# Example sentence
sentence = "This is an example sentence, and this sentence shows an example of word frequencies."

# Display word frequencies
frequencies = pandas_word_frequencies(sentence)
print(frequencies)
```

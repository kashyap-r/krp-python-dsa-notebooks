```python
#  My Logic 
a = 'Hi i am pawan here only only here i'
words = a.split()
print (f"List = {words}") 
uniqueWords = set(words)
print (f"Unique words = {uniqueWords}")
```

```python
a = 'Hi i am pawan here only only here i'
b = []

for word in a.split():
    if word not in b:
        b.append(word)
print (b)
```

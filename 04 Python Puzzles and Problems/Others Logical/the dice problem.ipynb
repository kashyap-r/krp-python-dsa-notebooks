## The Dice Problem 

You are given a cubic dice with 6 faces. All the individual faces have a number printed on them. THe numbers are in the range 1-6, like any ordinary dice. You will be provided the face of this cube, your task is to guess the number on the opposite face of the cube. 

Ex. Input: n= 2 => Output is 5 
Input: n = 6 => Output is 1 

```python
# Using if-else statement 

class theDiceProblem:
    def returnOppositeFace(self, n):
        if n == 1:
            return 6
        elif n == 2:
            return 5
        elif n == 3:
            return 4
        elif n == 4:
            return 3
        elif n == 5:
            return 2
        elif n == 6:
            return 1
        else:
            return 0 

if __name__ == "__main__":
    roll = theDiceProblem()
    n = 2
    result = roll.returnOppositeFace(n)
    if  result == 0:
        print ("Re-roll the dice")
    else:
        print ("The opposite face of the dice is ", result)
```

Time Complexity: O(1) i.e. the value in the variable is checked once
Space Complexity: O(1) i.e. uses one variable

```python
# Alternate method 
# Using Sum of two faces 
"""
The idea here is based on the observation that the sum of the two opposite
sides of a cubical dice is always equal to 7. 

i.e. 
1 is the opposite of 6
2 is the opposite of 5 
3 is the opposite of 4

Their respective sums are 7 

So, if the input face value is 2, then subtract 2 from 7 and we get 5 

"""

def returnOppface(n):
    return (7 - n)

n = 2
print (f"The opposite side of {n} is {returnOppface(n)}")

```

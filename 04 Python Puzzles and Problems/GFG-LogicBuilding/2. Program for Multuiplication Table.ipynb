```python
"""
Problem : Given a number print its multiplication table

Boundary conditions:
    1. Input Size (affects the complexity choice)
    2. Value Ranges (affects datatypes, memory)
    3. Special Cases (empty, inputs, duplicates, negative numbers)

"""

```

```python
# Iterative Approach

def printTable(num):
    for i in range (1, 11):
        print (f"{num}*{i} = {num*i}")

printTable(5)
```

```python
# recursive way of doing it
"""
In this method, we pass i as an additional parameter with initial value as 1. We print n * i and then recursively call for i+1. We stop the recursion when i becomes 11 as we need to print only 10 multiples of given number and i.
"""

def printTable2(num, i=1):
    if (i == 11):
        return
    print (num, "*", i, "=", num * i)
    i += 1
    printTable2(num, i)

if __name__ == "__main__":
    n = 5
    printTable2(n)
```

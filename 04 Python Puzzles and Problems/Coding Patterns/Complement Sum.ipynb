Complement Sum 

A complement sum problem typically refers to finding pairs of numbers in an array whose sum equals a given target value. This is a common interview question and can be approached in various ways.

```python
def complementSum(arr, target):
    seen = {}
    for i, num in enumerate(arr):
        complement = target - num 
        if complment in seen:
            return (seen[complement], i)
        seen[num] = i
    return None

arr = 
```

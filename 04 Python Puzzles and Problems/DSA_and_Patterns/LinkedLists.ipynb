```python
# Traversal of a singly linked list

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self):
        if self.head is None:
            self.head = Node(1)
            self.tail = self.head
        else:
            self.tail.next = Node(2)
            self.tail = self.tail.next

    def display(self):
        current = self.head
        while current is not None:
            print(current.data, end='->')
            current = current.next

    def printList(self):
        current = self.head
        while current is not None:
            print (current.data, end='->')
            current = current.next

if __name__ == "__main__":
    L1 = LinkedList()
    L1.append(1)
    L1.append(2)
    L1.append(3)

    L1.display()
    L1.printList()


```

### Python String Concatenation

    - allows to combine one or more strings into one
    - + operator 
    
ref: https://www.geeksforgeeks.org/python-string-concatenation/?ref=lbp

```python
s1 = 'Kashyap'
s2 = 'Raj'
fullname = s1 + ' ' + s2 + 'Purohit'
print (fullname)
```

### Using join method for concatenation 

Use join() function to concatenate strings with a specific separator. It’s especially useful when working with a sequence of strings, like a list or tuple. If no separator is needed then simply use join() with an empty string.

```python
str = ["Python", "is", "a", "crazy", "language", "for", "programming"]
sentence = " ".join(str)
print (sentence)

tup = ("Python", "is", "a", "super", "crazy", "language", "for", "programming")
now = " ".join(tup)
print (now)
```

Explanation: 

    ” “.join() method combines each element in words with a space between them.

    Note: join() is optimal for large-scale string concatenation since it minimizes the creation of intermediate strings.

### Using format() Method

    The format() method provided an easy way to combine and format strings. It is very helpful when we want to include multiple variables or values. 
    By using curly braces {} in a string, we can create placeholders that format() will fill in with the values we’ll provide.

https://www.geeksforgeeks.org/python-string-format-method/

```python
s1 = "Python"
s2 = "Programming"

res = "{} is a super crazy language for {}".format(s1, s2)
print (res)
```

### Using f-strings (Formatted Strings)

F-strings make it easy to combine and format strings in Python. By adding an f before the string, we can directly include variables in curly braces {}.

https://www.geeksforgeeks.org/formatted-string-literals-f-strings-python/

```python
s1 = "Python"
s2 = "Programming"

res = f"{s1} is a super crazy language for {s2}"
print (res)
```

### Which method to choose?

    + Operator: Simple and readable but not ideal for multiple concatenations in loops.
    
    join() Method: Highly efficient for concatenating lists of strings.
    
    f-strings: Use when variables are involved.
    
    format(): Similar to f-strings but compatible with earlier versions.

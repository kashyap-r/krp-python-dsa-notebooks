### Replace NONE by its previous NON None value
Find uncommon words from two strings
In this puzzle you have given two strings, you have to find uncommon words between two input strings

Input 
[None,1,2,3,None,4,None,None]

Output 
[None, 1, 2, 3, 3, 4, 4, 4]

```python
# Solution 1

ls = [None,2,3,None,None,5,None]
t = 0
rs = []
for i in ls:
    if i is not None:
        rs.append(i)
        t = i
    else:
        rs.append(t)
 
rs = [None if x == 0 else x for x in rs]
 
print(rs)
 
```

```python
# Solution #2
a = [None,2,3,None,None,5,None]
for i in range(0, len(a)):
    if a[i] == None and i != 0:
        for j in range(i-1, -1,-1):
            if a[j] != None:
                a[i] = a[j]
                break
print(a)
```

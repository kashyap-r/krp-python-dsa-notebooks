### Find all words from String present after given N words

Given a string S and a list lis[] of N number of words, the task is to find every possible (N+1)th word from string S such that the ‘second’ word comes immediately after the ‘first’ word, the ‘third’ word comes immediately after the ‘second’ word, the ‘fourth’ word comes immediately after the ‘third’ word and so on.

Boundaries:
1. min string length (input > srchLst)
2. whether comparison is case sensitive ?
3. what happens when input = srchLst
4. what happens when input != srchLst 

Input: S = “Siddhesh is a smart guy the city he lives in is a smart city”, lis: [“is”, “a”, “smart”]
Output: [guy, city]
Explanation: Here the two sequences where the words occur
just as the pattern mentioned are.

is a smart guy
is a smart city

Hence we found ‘guy’ and ‘city’ as the desired words. 


Input: S = “David loves to play and David loves to write sometimes”, lis: [“loves”, “to”]
Output: [play, write] 

#### Approach #1
Follow the steps to solve the problem:

1. Split the given string S by spaces.
2. Create a list ‘res‘ to store the ‘N+1’th words in the string.
3. Now, iterate through the split string from i = 0. 
   - Check every (i)th, (i+1)th, . . ., (i+N-1)th indices in every
   iteration, whether they are equal to the ‘first’, ‘second’, . . ., ‘N’th
   words respectively or not. 
   - If yes then append the (N+1)th word to the answer list.
4. Return the list storing the words as the required answer.

```python
str = 'Siddhesh is a smart guy the city he lives in is a smart city'
srchLst = ['is', 'a', 'smart']
words = str.split()

m = len(words)
n = len(srchLst)
for i in range(m-n):
    if ''.join(srchLst).lower() == ''.join(words[i:i+n]).lower():
        print (words[i+n])
```

Time Complexity : O(n * k * d) 
where k is size of search string and d is maximum size of the word

Auxiliary space: O(N)

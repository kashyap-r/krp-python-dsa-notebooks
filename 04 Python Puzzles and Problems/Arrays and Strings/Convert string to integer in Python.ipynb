### How to convert string to integer in Python?

https://www.geeksforgeeks.org/how-to-convert-string-to-integer-in-python/?ref=lbp



```python
# Method #1: Using built-in 
# int() - returns an integer value, equivalent of string in given base 

str = '100'
s2i = int(str)
print (s2i)
print (f"Conversion from {type(str)} to {type(s2i)}")
```

```python
str = '0x12F'
s2i = int(str, base=16)
print (s2i)
print (f"Conversion from {type(str)} to {type(s2i)}")
```

# 1. Using Regular Expressions

Explanation:
re.escape(article): Escapes any special characters in the article to ensure they are treated as literal strings.
\b: Word boundary anchor ensures that only whole words are matched.
re.IGNORECASE: Makes the search case-insensitive.



```python
# List of articles to search for
articles = ['a', 'an', 'the']

# Create a dictionary to store the counts
article_counts = {article: 0 for article in articles}

article_counts
```

```python
import re

text = "The quick brown fox jumps over a lazy dog. An apple a day keeps the doctor away."

# List of articles to search for
articles = ['a', 'an', 'the']

# Create a dictionary to store the counts (using dictionary comprehension)
article_counts = {article: 0 for article in articles}

# Search for each article and count occurrences
for article in articles:
    # Use \b to ensure we match whole words only
    pattern = r'\b' + re.escape(article) + r'\b'
    count = len(re.findall(pattern, text, re.IGNORECASE))
    article_counts[article] = count

print(article_counts)
```

# 2. Using str.split() and Manual Counting

If you want a more manual approach without regular expressions, you can split the text into words and count occurrences. This method also lets you handle punctuation manually:

```python
import string

text = "The quick brown fox jumps over a lazy dog. An apple a day keeps the doctor away."

# List of articles to search for
articles = ['a', 'an', 'the']

# Clean the text by removing punctuation and converting to lowercase
translator = str.maketrans('', '', string.punctuation)
cleaned_text = text.translate(translator).lower()

# Split the cleaned text into words
words = cleaned_text.split()

# Count occurrences
article_counts = {article: words.count(article) for article in articles}

print(article_counts)
```

**Explanation:**

    str.maketrans('', '', string.punctuation): Creates a translation table that removes punctuation.

    text.translate(translator).lower(): Cleans the text and converts it to lowercase.

    words.count(article): Counts occurrences of each article in the list.

**Summary**

    Regular Expressions: Ideal for precise matching of whole words with boundary checks and case insensitivity.
    
    Manual Counting with split(): Simple approach suitable for controlled text scenarios where punctuation handling is necessary.
    
    Both methods are effective, and the choice depends on the complexity of your text and whether you need case-insensitive matching or more flexibility with punctuation.

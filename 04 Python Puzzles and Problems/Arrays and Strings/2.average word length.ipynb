```python
import re

def average_word_length(sentence):
    # Extract words using regex
    words = re.findall(r'\w+', sentence)
    
    # Calculate total length of all words
    total_length = sum(len(word) for word in words)
    
    # Calculate average length
    avg_length = total_length / len(words) if words else 0
    
    return avg_length

# Example sentence
sentence = "This is an example sentence, and it calculates the average word length."

# Display average word length
avg_length = average_word_length(sentence)
print(f"Average word length: {avg_length:.2f}")
```

```python
import nltk

def nltk_average_word_length(sentence):
    # Tokenize the sentence into words
    words = nltk.word_tokenize(sentence)
    
    # Filter out punctuation
    words = [word for word in words if word.isalpha()]
    
    # Calculate total length and average word length
    total_length = sum(len(word) for word in words)
    avg_length = total_length / len(words) if words else 0
    
    return avg_length

# Example sentence
sentence = "This is an example sentence, and it calculates the average word length."

# Display average word length
avg_length = nltk_average_word_length(sentence)
print(f"Average word length: {avg_length:.2f}")
```

```python
import pandas as pd
import re

def pandas_average_word_length(sentence):
    # Extract words using regex
    words = re.findall(r'\w+', sentence)
    
    # Convert words to a pandas DataFrame
    df = pd.DataFrame(words, columns=['word'])
    
    # Calculate average word length
    avg_length = df['word'].apply(len).mean()
    
    return avg_length

# Example sentence
sentence = "This is an example sentence, and it calculates the average word length."

# Display average word length
avg_length = pandas_average_word_length(sentence)
print(f"Average word length: {avg_length:.2f}")
```

```python
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

def spacy_average_word_length(sentence):
    # Process the sentence using spaCy
    doc = nlp(sentence)
    
    # Filter out non-alphabetic tokens
    words = [token.text for token in doc if token.is_alpha]
    
    # Calculate total length and average word length
    total_length = sum(len(word) for word in words)
    avg_length = total_length / len(words) if words else 0
    
    return avg_length

# Example sentence
sentence = "This is an example sentence, and it calculates the average word length."

# Display average word length
avg_length = spacy_average_word_length(sentence)
print(f"Average word length: {avg_length:.2f}")
```

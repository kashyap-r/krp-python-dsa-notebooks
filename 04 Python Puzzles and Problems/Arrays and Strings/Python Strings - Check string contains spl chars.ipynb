Program to check if a string contains any special character

Given a string, the task is to check if that string contains any special character (defined special character set). If any special character is found, donâ€™t accept that string.


```python
# Approach #1: Using regular expression 

import re 

def run(string):
    # Make own character set and pass this as argument in compile method
    regex = re.compile('[@_!#$%^&*()<>?/\|}{~:]')

    if (regex.search(string) == None):
        print ("The string has no Spl. characters")
    else:
        print ("The string has spl. characters")


if __name__ == '__main__':
    str1 = 'Kashyap$Rajpurohit'
    run (str1)
    str2 = 'Kashyap Rajpurohit'
    run (str2)
```

```python
# Using builit-in methods
def has_special_char(s):
    for c in s:
        if not (c.isalpha() or c.isdigit() or c == ' '):
            return True
    return False


# Test the function
s = "Hello World"
if has_special_char(s):
    print("The string contains special characters.")
else:
    print("The string does not contain special characters.")

s = "Hello@World"
if has_special_char(s):
    print("The string contains special characters.")
else:
    print("The string does not contain special characters.")
```

The time complexity of this function is O(n), where n is the length of the input string, because it involves a single loop that iterates through all the characters in the string.

The space complexity of this function is O(1), because it does not use any additional data structures and the space it uses is independent of the input size.

```python
# Using String Punctuaation

import string 

def strCheck(s):
    for c in s:
        if c in string.punctuation:
            print ("String has spl. characters")
            return 
    print ("String does not have spl. characters")

strCheck("Kashyap%Rajpurohit")
strCheck("Kashyap Rajpurohit")
```

Time complexity: O(n), where n is the length of the input string. The loop iterates over each character in the string once.

Space complexity: O(1), as we are using only the string.punctuation string from the string module, which has a constant length.

# Using ASCII Values 

Approach 
1. Initialize for loop to iterate through each character of the input string 
2. for each character check its ASCII value
3. If ASCII value falls outseide the range, thn it is a Special Character

```python
def checkSplChar(str):
    for char in str:
        if ord(char) < 48 or (57 < ord(char) < 65) or (90 < ord(char) < 97) or ord(char) > 122:
            return "String has Spl. characters"
    return "String has No Spl. characters"

str1 = "Kashyap % Raj"
print (checkSplChar(str1))

print (checkSplChar('Kashyap'))
```

Time Complexity: The function iterates through each character in the string once, so the time complexity is O(n), where n is the length of the string.

Space Complexity: The function uses constant space, as it only creates a few variables to store the input string and the ASCII values of the characters. Therefore, the space complexity is O(1).

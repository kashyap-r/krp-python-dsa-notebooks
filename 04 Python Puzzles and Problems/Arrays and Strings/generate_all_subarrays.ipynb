Generate all Subarrays of an array

Task: Given an array, generate all possible subarrays of the given array

Ex. Input arr = [1, 2, 3]
Output = [[1], [1, 2], [2], [1, 2, 3], [2, 3], [3]]

```python
# Iterative Approach 
""" 
Have three loops
1. Outermost loop: Picks startig index of the current subarray 
2. Middle Loop: Picks the ending index of the current subarray
3. Innermost loop: Prints the sub array from starting index to the ending index 
"""

arr = [1, 2, 3]

n = len(arr)
# print (n)
# pick starting point
print (f"Input Array: [{arr}]")
print (f"All non empty sub arrays:\n")
for i in range (n):
    # pick ending point
    for j in range(i, n):
        #print the subarray between starting point and ending point        
        for k in range(i, j+1):
            print (f"{arr[k]}", end=" ")
        print () # new line after each sub array
```

Time Complexity: n*n*n
Space Complexity: o(1) - only one extrac location is needed

```python
# print sub arrays in O(n^2) 

arr = [1, 2, 3]
n = len(arr)
for i in range (n):
    for j in range (i, n):
        print (arr[i:j+1])
```

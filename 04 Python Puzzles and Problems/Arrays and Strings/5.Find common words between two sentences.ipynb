```python
# My logic 

a = 'Hi I am Pawan. I am here only'
b = 'here only'

d = {}

for word in a.split():
    word = word.strip(",.!?:;")
    if word in b:
        if word in d:
            d[word] += 1
        else:
            d[word] = 1
        

d
```

```python
#Find common words from two sentences
 
a = 'Hi I am Pawan. I am here only'
b = 'here only'
d = {}
 
for w1 in a.split():
    for w2 in b.split():
        if w1 == w2:
            d[w1] =  d.get(w1,0) + 1
 
print(d)
```

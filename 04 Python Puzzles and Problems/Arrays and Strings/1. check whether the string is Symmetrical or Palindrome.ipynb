### Python program to check whether the string is Symmetrical or Palindrome

Problem Description:
Given a string. the task is to check if the string is symmetrical and palindrome or not. 

A string is said to be symmetrical if both the halves of the string are the same and a string is said to be a palindrome string if one half of the string is the reverse of the other half or if a string appears the same when read forward or backward.

Analysis:
1. Find whether the string is symmetrical
2. Find whether the string is palindrome

**Approach 1 (Naive Approach using loops):**

For palindrome: 
    - use a loop to compare characters from the start to the end moving towards the middle

For Symmetry:
    - split the string into two and compare 


```python
def isPalindrome(str):
    revStr = ""
    for ch in str:
        revStr = ch + revStr        
    return str == revStr

def isSymmetrical(str):
    strlen = len(str)
    mid = strlen // 2
    if strlen % 2 == 0:
        return str[:mid] == str[mid:]
    else:
        return str[:mid] == str[mid+1:]

str = "amaaama"
str1 = "mama"
str2 = "mabma"
if isSymmetrical(str1):
    print ("The input string is symmetrical")
else: 
    print ("The input strig is not symmetrical")
if isPalindrome(str):
    print ("The input string is Palindrome")
else: 
    print ("The input strig is not Palindrome")
```

```python
Time Complxity: O(n)
Auxiliary Space: O(n)
Where n is number of characters in the string 
```

### Approach 2 - Using Slicing
We use slicing to reverse the string for palindrome checking and split the string for symmetry checking.


```python
string = "amaama"
mid = len(string) // 2
first_str = string[:mid]
second_str = string[mid:] if len(string) % 2 == 0 else string[mid+1:]

if string == string[::-1]:
    print ("is Palindrome!")
else:
    print ("is NOT Palindrome!")
if first_str == second_str:
    print ("is Symmetrical!")
else:
    print ("is NOT Symmetrical!")
```

Time Complxity: O(n)
Auxiliary Space: O(n)
Where n is number of characters in the string 

### Aproach 3 - Using reversed() function 

we can check the strings whether it is palindrome or not by using the reversed() function to the string which reverses the string which makes us easy to write the program.


```python
# for palindrome 
str = "smalayalams"
revStr = "".join(reversed(str))
if str == revStr:
    print ("is Palindrome!")
else:
    print ("is NOT Palindrome!")

# for symmetry.. just split the string into half
str1 = "madam"
mid = len(str1) // 2
first =  str1[:mid]
if len(str1) % 2 == 0:
    second = str1[mid:]
else:
    second = str1[mid+1:]
if first == second:
    print ("the string is symmetrical")
else:
    print ("the string is not symmetrical")
    
```

```python
The Time and Space Complexity for all the methods are the same:

Time Complexity: O(n)
Auxiliary Space: O(n)
```

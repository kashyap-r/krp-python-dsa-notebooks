### Python String Length 
- The string len() function returns the length of the string.

https://www.geeksforgeeks.org/python-string-length-len/?ref=lbp


```python
s1 = 'abcd'
print (len(s1))

s2 = ''
print (len(s2))
```

```python
tup = (1, 2, 3)
print (len(tup))

l = [1, 2, 3, 4]
print (len(l))
```

```python
# whre it fails 
print (len(True))
```

```python
dic = {'a':1, 'b':2}
print (len(dic))

s = {1, 2, 3, 4, 4}
print (len(s))
print (s)
```

```python
class Public:
    def __init__(self, number):
        self.number = number 
    def __len__(self):
        return self.number

obj = Public(21)
print (len(obj))
```

```python
l = range(1, 10)
len(l)
```

Does String len() Count Spaces?
Yes, the len() function counts spaces as characters when calculating the length of a string. Every character, including spaces, punctuation, special characters, and normal alphabetic or numeric characters, is counted as one towards the total length.

```python
# 1 Using Iteration
# split the string in substring manner

Input = 'Geeks_for_geeks_is_best'

splitStr = Input.split('_')
# print (splitStr)
output = []
for a in range (len(splitStr)):
    # print (splitStr[a])
    # print (splitStr[:a])
    temp = splitStr[:a+1]
    temp = '_'.join(temp)
    output.append(temp)

print (output)

"""
time COmplexity: O(1n) where n is the length of the string
space complexity: O(n) where n is the length of the string
"""
```

```python
# Using itertools
"""
Time complexity: O(n), where n is the length of the input string.

Auxiliary space: O(n), as we are creating a list to store the output of the accumulate function.
"""
from itertools import accumulate
input = 'Geeks_for_geeks_is_best'
output = [*accumulate(input.split('_'), lambda temp1, temp2: '_'.join([temp1, temp2])), ]
print (output)
```

```python
"""
Method 3: Using re module and string slicing

Time complexity: O(n), where n is the length of the input string.

Auxiliary space: O(m), where m is the number of substrings generated.
"""

import re
input = 'Geeks_for_geeks_is_best'

res = []
for i in re.finditer("(_)", input):
    temp = input[:i.span()[0]]
    res.append(temp)

res.append(input)
print (res)
```

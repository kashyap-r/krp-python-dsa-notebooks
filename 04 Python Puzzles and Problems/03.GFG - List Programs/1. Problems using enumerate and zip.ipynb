### Example 1: Compare Two Class Sections

Problem Statement:
You have two lists: one with student names from Section A and another with student names from Section B. You want to compare them pair‑wise and print a message showing whether the students at the same index are identical or different. Use enumerate() and zip() together.

SecA = ['Raj', 'Sheldon', 'Howard', 'Leonard', 'Comedy' ]

SecB = ['Ross', 'Joey', 'Chandler', 'Gunther', 'Comedy']



```python
SecA = ['Raj', 'Sheldon', 'Howard', 'Leonard', 'Comedy' ]

SecB = ['Ross', 'Joey', 'Chandler', 'Gunther', 'Comedy']

for idx, (a, b) in enumerate(zip(SecA, SecB), start=1):
    if a == b:
        print (f"{idx}. The pair {a} and {b} are a match")
    else:
        print (f"{idx}. The pair {a} and {b} are not a match")
```

### Example 2: Weighted Average of Marks

Problem Statement:

You have a list of student names, their marks, and a list of weight factors (importance of each exam). Compute the weighted score for each student and display it with an index.


```python
names = ["john", "ala", "ilia"]
marks = [80, 90, 70]
weights = [0.5, 0.3, 0.2]

for idx, (name, mark, wt) in enumerate(zip(names, marks,weights), start = 1):
    weighted_score = mark + wt
    print (f"{idx}. {name} -> mark={mark}, weight={wt}, weighted score={weighted_score}")
```

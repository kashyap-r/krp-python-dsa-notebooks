Suppose we have two matrices A and B.
A = [[1,2],[3,4]]
B = [[4,5],[6,7]]

then we get,
A+B = [[5,7],[9,11]]
A-B = [[-3,-3],[-3,-3]]

# 1. Adding elements of the Matrix

we have used np.add() method to add elements of two matrices. If shape of two arrays are not same, that is arr1.shape != arr2.shape, they must be broadcastable to a common shape (which may be the shape of one or the other)


```python
import numpy as np 

A = np.array([[1, 2],[3, 4]])
B = np.array([[5, 6],[7, 8]])

print (f"The first matrix:\n{A}")
print (f"The second matrix:\n{B}")

print (f"Adding the two matrices together, we get.. \n{np.add(A,B)}")
```

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[4, 5], [6, 7]])

print("Printing elements of first matrix")
print(A)
print("Printing elements of second matrix")
print(B)

# subtracting two matrix
print("Subtraction of two matrix")
print(np.subtract(A, B))
```

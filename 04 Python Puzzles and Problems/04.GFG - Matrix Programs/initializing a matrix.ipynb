```python
rows = 3
cols = 4 

matrix = [[0 for _ in range(cols)] for _ in range(rows)]

for row in matrix:
    print (row)

print ()
# print elements 
for row in matrix:
    for ele in row:
        print (ele, end = ", ")
    print ()
```

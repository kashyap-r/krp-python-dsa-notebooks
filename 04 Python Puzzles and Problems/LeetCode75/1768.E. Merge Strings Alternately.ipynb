```python
"""
Statement: You are given two strings word1 and word2. Merge the strings by adding the letters in alternating order, starting with word1. If a string is longer than the other, append the additional letters nto the end of the merged string.

Return the merged string.

Example 1:
Input: Word1 = "abc", Word2 = "pqr"
Output: "apbqcr"
Example 2:
word1 = "le"
word2 = "leetcode"
Output: "lleeetcode"

COnditions:

1. find the longer string

"""
```

```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        l1 = len(word1)
        l2 = len(word2)
        i = 0
        Output = ""
        while i < max(len(word1), len(word2)):
            if i < l1:
                Output += word1[i]
            if i < l2:
                Output += word2[i]
            i += 1
        return Output

sol1 = Solution()
str1 = "Kashyap"
str2 = "leetcode"
res = sol1.mergeAlternately(str1, str2)
print(res)


```

### While loop exercise question based on bouncing ball problem

```python
initial_height = 10
bounce_factor = 0.5
height = initial_height
while height > 0.1:
    print ("The ball is at a height of", height, "meter")
    height *= bounce_factor
print ("The ball has stopped bouncing.")
```

### Simple while-loop exercise code to build countdown clock

```python
countdown = 10 
while countdown > 0:
    print (countdown)
    countdown -= 1
print ("Blast off!")    
```

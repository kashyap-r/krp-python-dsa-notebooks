### Get the Size of a Dictionary

We are given a dictionary in Python and our task is to find the memory size that the dictionary occupies. This can be important when you need to know how much memory your data structures are consuming especially for large dictionaries.


```python
!pip install pympler
```

```python
# 1. Using sys.getsizeof()

import sys

d = {'apple': 5, 'carrot': 2, 'orange': 3, 'cherry': 9, 'peach': 4}
print(d)

size = sys.getsizeof(d)
print (f"Size of dictionary: {size} bytes")

# 2. Using __sizeof__() Method - Python also has an inbuilt __sizeof__() method to determine the space allocation of an object without any additional garbage value

size = d.__sizeof__()
print (f"Size of dictionary: {size} bytes")

# Using pympler.asizeof
from pympler import asizeof

size = asizeof.asizeof(d)
print (f"Size of dictionary: {size} bytes")
```

```python
d1 = {'a': 1, 'b': 2, 'c': {'nested_key': [1, 2, 3]}}
size = asizeof.asizeof(d1)
print (f"Size of dictionary: {size} bytes")
```

### Iterate over a dictionary in Python
1.
    - Iterate through keys
    - Iterate through values
    - Iterate through both keys and values

2. Iterating Python Dictionary Using map() and dict.get
3. Iterate Python Dictionary using zip() Function
4. Dictionary iteration in Python by unpacking the dictionary

```python
d = {"name": 'John', "age": 99, 'location': "B'luru"}
print (d)
```

```python
# Iterate through Keys
for key in d:
    print (key)
print()
for key in d.keys():
    print(key)
```

```python
# Iterate through values
for value in d.values():
    print(value)
```

```python
# Iterate through both keys and values
for key, value in d.items():
    print(key, value)
print()
for key, value in d.items():
    print(f"{key}: {value}")
```

```python
# Iterating Python Dictionary Using map() and dict.get

statesAndCapitals = {
    'assam': 'dispur',
    'bihar': 'patna',
    'tamilnadu': 'chennai',
    'karnataka': 'bengaluru',
    'gujarat': 'gandhinagar',
    'maharastra': 'mumbai'
}

map_keys = map(statesAndCapitals.get, statesAndCapitals)
print (map_keys)
for key in map_keys:
    print(key)
```

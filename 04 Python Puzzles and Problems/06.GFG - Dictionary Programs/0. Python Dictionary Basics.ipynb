```python
# Creating a dictionary using {}

d = {'kash':10, 'raj': 20, 'purohit': 15}

d1 = {1:'Kash', 2:'raj', 3:'purohit'}

print(d)
print (d1)

# creating a dictionary using a constructor
d2 = dict(a='KRP', b='UK', c='SR')
print (d2)
```

- Dictionary keys are case-sensitive 
- Keys must be immutable i.e. keys can be strings, numbers or tuples
- keys must be unique
- Dictionary internally uses Hashing. Hence, operations like search, insert, delete can be performed in Constant Time. 


```python
d = {'kash':10, 'kash': 20, 'purohit': 15}

print (d)
# Does not error out but is not considered

d = {'kash':10, 'Kash': 20, 'purohit': 15}
print (d)
# Dictionary keys are case sensitive
```

```python
# Example of a dictionary where key is a tuple
d = {(1, 'kash'):'Male', 
     (2, 'Shreya'): 'Female', 
     (3, 'Usha'): 'Female'}

print (d)
```

```python
# Example dictionary with tuple as key
tuple_dict = {
    (1, 'apple'): 'fruit',
    (2, 'carrot'): 'vegetable',
    (3, 'chicken'): 'meat'
}

# Accessing values using tuple keys
print(tuple_dict[(1, 'apple')])    # Output: fruit
print(tuple_dict[(2, 'carrot')])   # Output: vegetable
print(tuple_dict[(3, 'chicken')])  # Output: meat
```

# Removing Dictionary Items


del : removes an item by key 
pop(): removes and item by key and returns its value 
clear(): empties the dictionary 
popitem(): removes and returns the last key-value pair

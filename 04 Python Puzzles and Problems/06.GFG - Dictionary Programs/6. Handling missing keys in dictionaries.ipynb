
### Handling missing keys in python dictionaries 


```python
dic = {'a': 1, 'b': 2}

print (dic)
print (dic['c'])
```

```python
# Using defaultdict 
from collections import defaultdict 

dic = defaultdict (lambda: 'Key not found')

dic['a'] = 1
dic['b'] = 2

print (dic['a'])
print (dic['c'])
```

```python
# Using the get() method 
dic= {'India': '0091', 'Australia': '0025', 'Nepal': '00977'}

print(dic.get('India', 'Not Found'))  
print(dic.get('Japan', 'Not Found'))
```

```python
# Using a setdefault() Method 

d = {'India': '0091', 'Australia': '0025', 'Nepal': '00977'}

d.setdefault('Japan', 'ISD Code Not Present')
print (d['India'])
print (d['Japan'])
```

```python
# Choosing dict.get() Over dict[] in Python

# 1. Handling Missing keys 
d = {'a': 1, 'b': 2}

val  = d['c']

print (val)
```

```python
d = {'a': 1, 'b': 2}

val  = d.get('c')
# using the get method does not result in an error 
print (val)

# alternatively provide default return values 
val = d.get('c', 0)
print (val)
```

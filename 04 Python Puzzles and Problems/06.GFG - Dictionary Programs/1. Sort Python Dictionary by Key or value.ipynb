```python
# 1. Sorting by values - using sorted() with lamda
d = {'apple': 5, 'banana': 2, 'cherry': 7}
asc = {k: v for k, v in sorted(d.items(), key=lambda item: item[1])}
print (asc)
```

```python
#2.. Using ordered dict

from collections import OrderedDict
d = {'apple': 5, 'banana': 2, 'cherry': 7}
res = OrderedDict(sorted(d.items(), key=lambda item: item[1]))
print (res)
```

```python
# lambda

# Add 10 to a number
add10 = lambda a: a + 10
print(add10(5))   # Output: 15

# Multiply two numbers
multiply = lambda a, b: a * b
print(multiply(3, 4))   # Output: 12

# Sort a list of tuples by the second element
pairs = [(5, 'orange'), (1, 'apple'), (2, 'banana'), (3, 'cherry')]
sorted_pairs = sorted(pairs, key=lambda x: x[1])
print(sorted_pairs)   # Output: [(1, 'apple'), (2, 'banana'), (3, 'cherry')]

```

#3 Using for loop with sorted

```python
d = {'apple': 5, 'banana': 2, 'cherry': 7, 'mango': 1}

for k, v in sorted(d.items(), key=lambda item: item[1]):
    print ((k, v), end = ' ')
```

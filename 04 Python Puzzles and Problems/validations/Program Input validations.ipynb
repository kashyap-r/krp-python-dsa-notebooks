**1. Validations for Numeric Values** 

- Check if NUmeric: Whether a value is a number (integer or a float)
- Range Check: Verify that the numberic values falls within a specified range
- Precision Check: Ensure the number has a certain number of decimal places

```python
# Check if an input value is numeric and non-negative

def validateNumeric(value):
    try:
        num = float(value) # check if numeric 
        if 0 <= num <= 100: # Range check
            print ("Valid Numeric Input")
        else:
            print ("Invalid Numeric Input")
    except ValueError:
        print ("Invalid NUmeric Input")

validateNumeric(10.5)
validateNumeric(-10)
validateNumeric('abc')  
```


2. Validation for Integer Values
- Check if integer 
- Range check 
- Odd / Even Check 

```python
def validateInteger(value):
    if isinstance(value, int):
        if value % 2 == 0:
            print ("Value is Integer and even")
        else:
            print ("value is integer and odd")
    else: 
        print ("value is not an integer")

validateInteger(10.0)
# validateInteger(abc)
validateInteger('abc')
validateInteger(-10)
validateInteger(5)
```

3. Validations of String values

Length Check:  Ensure the string is of certain length  
Patter Check: Validate the string against a regex pattern (email format) 
Non-Empty Check: Ensure the string is not empty 
Character Set check: Ensure the string contains only specific characters

```python
import re 
def validateString(value):
    if isinstance(value, str):
        if (len(value)>0):
            if re.match(r'^[a-zA-Z]+$', value):
                print ("Valid Alphabetic string")
            else:
                print ("String contains non alphabetic characters")
        else:
            print ("String is empty!!")
    else:
        print ("Invalid String Input")

validateString("Hello")
validateString("A")
validateString("$")
validateString("")
validateString(" ")
validateString(12)
validateString("123")
```

4. Validations for Character Values

- Single Character Check 
- Alphabet Check 
- Digit Check
- Special Character Check


```python

def validateCharacter(value):
    # single character check 
    if isinstance(value, str) and len(value) == 1:
        if value.isalpha():
            print ("Valid alphabet")
        elif value.isdigit():
            print ("Valid digit")
        else:
            print ("special character")
    else:
        print ("invalid character input")

validateCharacter('A')
validateCharacter('d')
validateCharacter('12')
validateCharacter('4')
validateCharacter('9')
validateCharacter(123)
validateCharacter(' ')
validateCharacter('$')
```

```python
# 1. Using a loop and type checking 
# This approach iterates through each element in the array and cheks if it is numeric

def checkArray(arr):
    for ele in arr:
        if not isinstance(ele, (int, float)) or ele < 0:
            return False

    return True 

arr = [10, 5.5, 3. -1, 'hello']
arr = [10, 5.5, 3. -1]
if checkArray(arr):
    print ("Only Numeric")
else:
    print ("non numberic")
```

```python
# Using list comprehension 
#  che k for all elements in the array being numeric and positive
def checkArray(arr):
    return all(isinstance(ele, (int, float)) and ele >= 0 for ele in arr)

arr = [10, 1, 2, 3.5]
if checkArray(arr):
    print ("Good Array ")
else:
    print ("bad array")
```

```python
# Using try-except for conversion and checking 

def check_array(arr):
    for element in arr:
        try:
            # Attempt to convert to float and check if non-negative
            if float(element) < 0:
                return False
        except ValueError:
            return False
    return True

array = [10, 5.5, 3, -1, 'hello']  # Example array
if check_array(array):
    print("The array has only numeric and non-negative elements.")
else:
    print("The array has non-numeric or negative elements.")


```


Validation on Arrays

    1. Length Validation 
    2. Element Type Validation 
    3. Range Validation 
    4. Uniqueness Validation 
    5. Non-empty Validation 
    6. Element value validation 
    7. Sorted Validation 
    8. Nester Array Validation
    9. Custom Rule Validation 

```python
# 1. Length Validation - Ensure the array has reqd number of elements 
def validateArrayLength(arr, minLen):    
    if not len(arr) > minLen:
        return 0 
    else:
        return 1

res = validateArrayLength([10, 12, 56], 2)
print (f"Valid array = {'True' if res else 'False'}")
```

```python
# 2. Element type validation - ensure all the elements in the array are of a specific type (ex. integers)

def validateArrayType(arr, elementType):
    if not all(isinstance(ele, elementType) for ele in arr):
        print (f"array has elements not of type {elementType}!! ")
        return 0
    else:
        print (f"All elements in array Are of type {elementType}! ")
        return 1

arr = [1, 2, 3, 5]
# elementType = int
res = validateArrayType(arr, int)
print (f"Is this an array of integers ? {'True' if res else 'False'} ")
```

```python
# 3. Range Validation: Ensure all numeric elements in the array fall within a specified range

def validateArrayRange(arr, minVal, maxVal):
    if all(minVal <= ele <= maxVal for ele in arr if isinstance(ele, (int, float))):
        print (f"All elements of array within range")
    else:
        print (f"Elements of array out of range")

validateArrayRange([10, 20, 30], 0, 200)
validateArrayRange([10, 20, 30], 0, 25)
```

```python
# 4. Uniqueness Validation - Ensure all elements in the array are unique

def validateUniqueness(arr):
    if len(arr) == len(set(arr)):
        print ("All elements are unique")
    else:
        print ("There are duplicate elements")

validateUniqueness([1, 2, 3, 4])
validateUniqueness([1, 2, 3, 3])
```

```python
# 5. Non-Empty Validation - Check if array is empty 

def validateNonEmpty(arr):
    if len(arr) > 0:
        print ("array is not empty")
    else:
        print ("array is empty")

arr = []
print (validateNonEmpty(arr))        
```

```python
# 6. Element Value Validation - Validate that all elements meet specific criteria (e.g., positive numbers, non-empty strings).

def validateArray(arr):
    if all(ele > 0 for ele in arr if isinstance(ele, (int, float))):
        print (f"All element are positive")
    else:
        print (f"some elements are not positive")

validateArray([1, 2, 4])
validateArray([-1, 0, 0])
```

```python
# 7. Sorted Validation - Ensure the array is sorted (ascending).
# arr = [1, 2, 3, 4]
arr = [4, 2, 1, 0]
arr == sorted(arr)
```

```python
# 8. Nested Array Validation - Purpose: Ensure that all elements in a multi-dimensional array are arrays themselves and validate their contents.
# Nested Array validation 

def validateNestedArray(arr):
    if all(isinstance(sublist, list) and len(sublist) > 0 for sublist in arr):
        print ("All elements are non-emptylists") 
    else:
        print ("Some elements are not non-empty sublists")

validateNestedArray([[1, 2],[3, 3],[4]])
validateNestedArray([[1, 2],[3, 3],4])
```

```python
# 9. Custom Rule Validation - Implement any custom rule specific to the application (e.g., certain elements must sum to a specific value).
# Check if the sum of elements is equal to a target value ... 

def validateSum(arr, target):
    if sum(arr) == target:
        print (f"The sum of elements is correct.")
    else:
        print ("The Sum of the elements is incorrect")

validateSum([1, 2, 3, 4], 10)
```

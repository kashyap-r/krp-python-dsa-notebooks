```python
""" 
Reverse a number digit by digit
"""

n = 4562
print (n)
rev = 0 

while n > 0:
    a = n % 10 
    rev = rev * 10 + a
    n = n//10

print (f"Reverse of {n} is {rev}")
```

```python
""" Using recursion """

def reverseDigits(n, revNum, basePos):
    if n > 0:
        reverseDigits(n // 10, revNum, basePos)
        revNum[0] += (n % 10) * basePos[0]
        basePos[0] *= 10



n = 4562 
revNum = [0] 
basePos = [1]
reverseDigits(n, revNum, basePos)
print (revNum[0])
```

```python
""" Using String """

n = 4562

s = str(n)

s = list(s)
s.reverse()
s = "".join(s)
s = int(s)
print (s)
```

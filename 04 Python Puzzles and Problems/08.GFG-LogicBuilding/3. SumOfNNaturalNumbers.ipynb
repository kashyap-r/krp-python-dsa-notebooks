```python
"""
Program for sum of n natural numbers

Given a positive integer n, find the sum of first n numbers

Questions:
1. Size of the input
2. Validations: what if the input is a 0 or a -ve number or a decimal
3. Output format
"""

n = 5
sum = 0 + 1+ 2 + 3 + 4
```

```python

def sumN(num):
    sum = 0
    for i in range(n):
        sum = sum + i
    return sum

if __name__ == '__main__':
    n = 6
    print(sumN(n))

"""
Time: O(n)
Space: O(1)
"""
```

```python
#Using recursion

def findSum1(n):
    #base case
    if n == 0:
        return 0
    return n + findSum1(n - 1)

print(findSum1(6))
```

```python
def findSum(n):
    # base case
    if n == 1:
        return 1

    return n + findSum(n - 1)

if __name__ == "__main__":
    n = 6
    sum = 0
    sum = findSum(n)
    print(sum)
```

```python
# 3rd Method using the formula n(n+1) / 2

def findSum(n):

    # Using mathematical formula to compute
    # sum of first n natural numbers
    return n * (n + 1) // 2

if __name__ == "__main__":
    n = 6
    print(findSum(n))
```

```python
"""
The Dice Problem

You are given a cubic dice with 6 faces. All the individual faces haave a number printed on them.
The numbers are in the range 1 to 6. You will ve provided a face of this cube, your task is to guess the number on the opposite face of the cube

Face shown	Opposite face
1	6
2	5
3	4

So if the input is 1 output is 6
Input is 2 output is 5
"""
```

```python
# Approach #1 (Naive Approach) - using if then else statement
"""
Time Complexity: O(1)
Auxiliary Space: O(1)
"""
def oppFaceOfDice(n):
    if n == 1:
        return 6
    elif n == 2:
        return 5
    elif n == 3:
        return 4
    elif n == 4:
        return 3
    elif n == 5:
        return 2
    elif n == 6:
        return 1

if __name__ == '__main__':
    print(oppFaceOfDice(6))
    print(oppFaceOfDice(5))
```

```python
# Approach #2 - Using Sum of two sides
"""
Time Complexity: O(1)
Auxiliary Space: O(1)
"""
def retOppSideOfDice(n):
    return 7 - n

print(retOppSideOfDice(6))
print(retOppSideOfDice(5))
print(retOppSideOfDice(2))
```

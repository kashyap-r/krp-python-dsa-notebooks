```python
num = 5

fact = 1 
for i in range(1, num + 1):
    fact *= i 

print (f"The factorial of {num} is {fact}")
```

```python
# factorial using recursion 

def factorial(n):
    return 1 if (n == 1 or n == 0) else n * factorial(n - 1)

num = 5 
print (f"Factorial of {num} is {factorial(num)}")
```

```python
# using math library 
import math 

def factorial(n):
    return math.factorial(n)

    
print (f"Factorial of {num} is {factorial(num)}")
```

```python
# Find factorial of a number using Numpy 

import numpy 
n = 5 
fact = numpy.prod([i for i in range(1, n+1)])
print (f"Factorial of {n} is {fact}")
```

# Prime factoorization method of finding factorial 

1. Initialize the factorial variable to 1
2. For each number i from 2 to n, do the following:
   a. Find the prime factorization of i
   b. For each prime factor p and its corresponding power k in the factorizaion of i, multiply the factorial variable by p raised to the power of k
3. Return the factorial variable.


```python
# Function to find prime factors of a number
def primeFactors(n):
    factors = {}
    i = 2
    while i*i <= n:
        while n % i == 0:
            if i not in factors:
                factors[i] = 0
            factors[i] += 1
            n //= i
        i += 1
    if n > 1:
        if n not in factors:
            factors[n] = 0
        factors[n] += 1
    return factors

# Function to find factorial of a number
def factorial(n):
    result = 1
    for i in range(2, n+1):
        factors = primeFactors(i)
        for p in factors:
            result *= p ** factors[p]
    return result

# Driver Code
num = 5
print("Factorial of", num, "is", factorial(num))
```

Time Complexity: O(sqrt(n))
Auxiliary Space: O(sqrt(n))

There are many methods to add two number in Python:

1. Using “+” operator
2. Defining a function that adds two number
3. Using operator.add method
4. Using lambda function
5. Using Recursive function

Reference:
1. https://www.geeksforgeeks.org/python-program-to-add-two-numbers/?ref=lbp
   

```python
# 1. Add two numbers with '+' operator 

num1 = 10 
num2 = 12 

sum = num1 + num2 
print ("Sum of ", num1, "and", num2, "is", sum) 
```

```python
# Add 2 numbers with user input 

# Python3 program to add two numbers

number1 = input("First number: ")
number2 = input("\nSecond number: ")

# Adding two numbers

# User might also enter float numbers
sum = float(number1) + float(number2)

# Display the sum
# will print value in float
print("The sum of {0} and {1} is {2}" .format(number1, 
                                            number2, sum))
```

```python
# Add Two Numbers in Python Using Function

def addNum(a, b):
    return a + b

num1 = 12
num2 = 13
sum = addNum(num1, num2)
print (sum)
```

```python
# Using operator function 
import operator 

num1 = 12
num2 = 13

sum = operator.add(num1, num2)
print (sum)
```

```python
# Using Lambda function 
addNumbers = lambda x, y: x + y

num1 = 3
num2 = 5

result = addNumbers(num1, num2)
print (result)
```

```python
# Add two numbers using recursibe function 

def addNums(x, y):
    if y == 0:
        return x
    else:
        return addNums(x+1, y-1)

num1 = 11
num2 = 22

# Call the recursive function to add the two numbers
result = addNums(num1, num2)

# Print the result
print("The sum of", num1, "and", num2, "is", result)
```

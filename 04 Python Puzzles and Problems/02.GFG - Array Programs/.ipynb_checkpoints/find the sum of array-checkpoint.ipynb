Given an array of integers, find the sum of its elements.

Input : arr[] = {1, 2, 3}
Output : 6
Explanation: 1 + 2 + 3 = 6


```python
def sumArr(arr):
    sum = 0
    for i in arr:
        sum = sum + i 
    return sum 

arr = [1, 2, 3, 4]
ans = sumArr(arr)
print (ans)
```

Time complexity: O(n), 
Auxiliary Space: O(1)

### Python Program to Find Sum of Array Using reduce method

Using the reduce method. Array.reduce() method is used to iterate over the array and get the summarized result from all elements of array.

```python
from functools import reduce 

def _sum(arr):
    # iterate over the array using reduce and get sum on accumulator
    sum = reduce(lambda a, b: a+b, arr)
    return (sum)

arr = [12, 3, 4, 15]
ans = _sum(arr)
 
# display sum
print('Sum of the array is ', ans)
```

### Sum of the array using the enumerate function 

This code calculates the sum of elements in the list list1 using a loop. It iterates through each element, adds it to the running sum s, and then prints the final sum.

```python
arr = [1, 2, 3, 4, 5]
sum = 0
for index, val in enumerate(arr):
    sum += val
print (sum)
```

```python
Time complexity: O(n)
Auxiliary Space: O(1)
```

# find sum of arrays using divide and conquer method 

- Define a function that takes an array and two indeces low and high
- if low == high, return the element at that index
- calculate the midpoint of the array as mid = (low + high) // 2
- recursively find the sum of the left subarray ans left_sum = sum_of_array(arr, low, mid)
- recursively find the sum of the right sub-array as right_sum = sum_of_array(arr, mid+1, high)
- return the sum of the left and right subarrays as left_sum + right_sum


```python
def sum_of_array(arr, low, high):
    if low == high:
        return arr[low]

    mid = (low + high) // 2 
    left_sum = sum_of_array(arr, low, mid)
    right_sum = sum_of_array(arr, mid+1, high)
    return left_sum + right_sum 

arr = [1, 2, 3, 4, 5]
print (sum_of_array(arr, 0, len(arr) - 1))

arr = [1, 2, 3]
print(sum_of_array(arr, 0, len(arr)-1)) # Output: 6
 
arr = [15, 12, 13, 10]
print(sum_of_array(arr, 0, len(arr)-1)) # Output: 50
```

### Program to find the sum of array using counter method 

This program finds the sum of an array using the Counter class from the collections module in Python. 

The Counter class is used to count the occurrences of elements in the input array. The program then iterates over the items in the counter and calculates the sum of the array.


```python
from collections import Counter 

arr = [1, 2, 3, 4, 5]
c = Counter(arr)
sum = 0 
for key, value in c.items():
    sum += key * value

print (sum)
```

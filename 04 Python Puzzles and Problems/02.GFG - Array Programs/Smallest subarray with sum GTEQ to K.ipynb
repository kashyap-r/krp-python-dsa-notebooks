Smallest subarray with sum greater than or equal to K

Problem Description
Given an array A[] consisting of N integers and an integer K, the task is to find the length of the smallest subarray with sum greater than or equal to K. If no such subarray exists, print -1. 

Input: A[] = {2, -1, 2}, K = 3 
Output: 3 
Explanation: 
Sum of the given array is 3. 
Hence, the smallest possible subarray satisfying the required condition is the entire array. 
Therefore, the length is 3. 

Constraints and COnsiderations
1. Array should have atleast 2 elements 
2. Array should have numeric elements 
3. Edge Cases - Solution should handle 
    - Negative elements in the array 
    - All identical elements in array 
    - All elements Zero in the array 


Naive Approach 
- Simplest approach is to generate all possible subarrays of the given array 
- Then check which subarray sum is greater than or equal to K
- Print the subarray having inimum length that satisfies 


```python
# function to find smallest sub array with sum GTEQ to K 
def findSubarray(arr, n, k):
    ans = float('inf')
    for i in range(n):
        # for sum of all elements of subarray 
        sum = 0 
        count = 0 
        for j in range(i, n):
            count += 1 
            sum += arr[j]
            if sum >= k:
                if count < ans:
                    ans = count
    # when no such subarray exists
    if ans == float('inf'):
        return -1 
    return ans

arr = [2, 1, 1, -4, 3, 1, -1, 2]
k = 5 
print (findSubarray(arr, len(arr), k))
```

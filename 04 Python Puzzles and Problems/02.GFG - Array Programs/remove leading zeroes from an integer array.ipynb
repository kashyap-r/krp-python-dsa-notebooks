<!-- Remove Leading Zeroes from an integer array  -->

Handling leading Zeroes in an integer array depdens on what you want to achieve. 
Below are few scenarios and their corresponding solution. 


```python
"""
Scenario 1: Treat the array as a number and remove leading zeroes 

- By converting the array into a string and then converting it back to an integer
"""

def removeLeadingZeroes(arr):
    # join the array into a string
    numStr = ''.join(map(str, arr))
    num = int(numStr)
    return (list(map(int, str(num))))

arr = [0, 0, 0, 1, 2, 3]
print ("Input :", arr)
result = removeLeadingZeroes(arr)
print ("Result: ", result)
```

```python
"""
Scenario #2: Removing the leading Zeroes and keeping the array intact 

"""

def removeLeadingZeroes(arr):
    index = 0 
    while index < len(arr) and arr[index] == 0:
        index += 1
    return arr[index:]

arr = [0, 0, 1, 2, 3]
result = removeLeadingZeroes(arr)
print (result)

```

```python
"""
Scenario #3: Keeping a Single Zero for all Zero Arrays 
If you want to keep a single zero when the array consisits only of zeros
"""
def removeLeadingZeroes(arr):
    index = 0 
    while index < len(arr) and arr[index] == 0:
        index += 1
    #  is all the elemnts in the array are zero
    if index == len(arr):
        return [0]
    return arr[index:]

arr = [0, 0, 0]
res = removeLeadingZeroes(arr)
print (res)

arr = [0, 0, 0.0, 1, 1, 2, 3, 0, 5]
res = removeLeadingZeroes(arr)
print (res)
```

```python
"""
Scenario #4: 
"""

def removeLeadingZeros(arr):
    return [int(str(x)) for x in arr]

arr = [09, 3, 54, 03, 33]
res = removeLeadingZeros(arr)
print (res)

```

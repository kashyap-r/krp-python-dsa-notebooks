### 238. Product of Array Except Self

Given an integer array nums, return an array answer such that answer[i] is equal to the product of all the elements of nums except nums[i].

The product of any prefix or suffix of nums is guaranteed to fit in a 32-bit integer.

You must write an algorithm that runs in O(n) time and without using the division opera


Example 1:

Input: nums = [1,2,3,4]
Output: [24,12,8,6]

Example 2:

Input: nums = [-1,1,0,-3,3]
Output: [0,0,9,0,0]
tion.

```python
# input = [1, 2, 3, 4]
input = [-1,1,0,-3,3]
arrlen = len (input)
prod = [1 for i in range(len(input))]
i_cnt = 0 
j_cnt = 0 
for i in range(arrlen):
    print (input[i])
    i_cnt += 1
    curr = input[i]
    for j in range(arrlen):
        j_cnt += 1
        if curr != input[j]:
            prod[i] *= input[j]

print (f"Output array: {prod}")
print (f"i_count : {i_cnt}")
print (f"j_count : {j_cnt}")
```

```python
def product_except_self(nums):
    n = len(nums)
    
    # Initialize the result array
    result = [1] * n
    
    # Compute prefix products
    prefix_product = 1
    for i in range(n):
        result[i] = prefix_product
        prefix_product *= nums[i]

    print (result)
    print (prefix_product)
    # Compute suffix products and finalize result
    suffix_product = 1
    for i in range(n - 1, -1, -1):
        result[i] *= suffix_product
        suffix_product *= nums[i]
    print (result)
    print (suffix_product)
    return result

# Example usage:
nums = [1, 2, 3, 4]
print(product_except_self(nums))  # Output: [24, 12, 8, 6]
```

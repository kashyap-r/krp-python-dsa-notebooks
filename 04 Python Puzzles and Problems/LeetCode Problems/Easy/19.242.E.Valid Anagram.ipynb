Question: What are the rules to follow to define an "anagram"?


Notes: Rules to form anagrams
1. Same Letters: Anagram must use exactly the same letters as the original word or Phrase.
   No additional letters can be added, none can be left out.
2. Same frequency of each letter
3. Rearrangement of letters -
4. Meaning (Optional) - Strictly speaking, an anagram does not need to form a meaningful word or phrase, but in many word puzzles or games, the goal is to form valid, meaningful words.
5. Case Insensitive - Anagrams are case insensitive, meaning uppercase and lowercase letters are treated the same.
6. Spaces and Punctuation (Varies) - In some cases, spaces and punctuation marks can be ignored when forming anagrams, depending on the context (such as word puzzles or games). For example, when forming anagrams of phrases, spaces are often ignored.
7. Length of Words - Both the original word and its anagram must have the same number of letters.
8. Partial Anagrams: where the length of the anagrams are lesser than the original word.

```python
from collections import defaultdict

class Solution:
    def validAnagram(self, str1, str2):
        count = defaultdict(int)        
        for chr in str1:
            count[chr] += 1               
        for chr in str2:
            count[chr] -= 1        
        for val in count.values():
            if val != 0:
                return False
        return True

soln = Solution()
str1 = 'kashyap1#'
str2 = 'yapkash1#$'
print (soln.validAnagram(str1, str2))
```

### Method 1: Using a Set (Efficient)

Time COmplexity: O(n), because we scan thru the array once 
Space Complexit: O(n), because we use a set to store the existing elements


```python

def containsDuplicate(arr):
    seen = set()
    for num in arr:
        if num in seen:
            return True
        seen.add(num)
    return False
        
arr = [1, 2, 3, 4, 2]
print("True" if (containsDuplicate(arr) > 0) else 'False')
```

```python
# Hash Map
"""
In this approach, we iterate through the array and store each element as a key in a hash map. The value associated with each key represents the count of occurrences of that element. If we encounter an element that already exists in the hash map with a count greater than or equal to 1, we return true, indicating that a duplicate has been found. Otherwise, we update the count of that element in the hash map. If we complete the iteration without finding any duplicates, we return false.
"""

class Solution:
    def containsDuplicate(self, nums) -> bool:
        seen = {}
        for num in nums:
            if num in seen and seen[num] >= 1:
                return True
            seen[num] = seen.get(num, 0) + 1
        return False
        
sln = Solution()
arr = [1, 2, 3, 4, 12]
print("True" if (sln.containsDuplicate(arr) > 0) else 'False')
```

### Method 2: Using Sorting (Less Efficient in Space)

Sort the array first, then check for consecutive duplicate elements
This modifies the original array 

Time Complexity: ***O(n log n)*** due to sorting
Space compelxity: O(1) if you modify the original array, O(n) if you create a copy.

```python
def containsDuplicate(arr):
    arr.sort()
    for i in range(1, len(arr)):
        if arr[i] == arr[i-1]:
            return True
    return False

arr = [1, 2, 3, 4, 2]
print("True" if (containsDuplicate(arr) > 0) else 'False')
```

### Method #3: Using collections.Counter (Finding all duplicates) 


```python
from collections import Counter

def containsAllDuplicates(arr):
    duplicates = -1
    counter = Counter(arr)
    duplicates = [num for num, count in counter.items() if count > 1]
    return duplicates

arr = [1, 2, 3, 4, 5, 6, 1]
duplicates = containsAllDuplicates(arr)
print (duplicates)
```

### Method 4: Using Floyd’s Tortoise and Hare (Cycle Detection) — When only one duplicate exists

```python
def find_duplicate(arr):
    slow = arr[0]
    fast = arr[arr[0]]

    # Phase 1: Find intersection point
    while slow != fast:
        slow = arr[slow]
        fast = arr[arr[fast]]

    # Phase 2: Find the entrance to the cycle (duplicate)
    slow = 0
    while slow != fast:
        slow = arr[slow]
        fast = arr[fast]

    return slow

# Example usage:
arr = [1, 3, 4, 2, 5, 3]
duplicate = find_duplicate(arr)
print("Duplicate element:", duplicate)
```

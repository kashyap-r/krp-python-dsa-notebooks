```python
# Given array 
arr = [1, 2, 3, 4, 5, 3, 0, 6]
# to delete all occurances of below value
item_to_delete = 3 
# output 
# return the remaining sorted array and arraycount 

arr = [item for item in arr if item != item_to_delete]
n = len(arr)
for i in range(n-1):
    for j in range(i+1, n):
        if arr[i] > arr[j]:
            arr[i], arr[j] = arr[j],arr[i]
            
print (arr)

```

```python
class Solution:
    def removeElement(self, nums, val) -> int:
        nums = [item for item in nums if item != val]
        k = len(nums)
        for i in range(k-1):
            for j in range(i+1, k):
                if nums[i] > nums[j]:
                    nums[i], nums[j] = nums[j], nums[i]
        return nums, k

soln = Solution()
nums = [3,2,2,3]
val = 3
print (soln.removeElement(nums, val))
```

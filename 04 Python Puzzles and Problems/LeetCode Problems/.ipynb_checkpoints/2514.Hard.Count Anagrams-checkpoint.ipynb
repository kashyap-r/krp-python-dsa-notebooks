2514. Count Anagrams

You are given a string s containing one or more words. Every consecutive pair of words is separated by a single space ' '

A string t is an anagram of string s if the ith word of t is a permutation of the ith word of s.

    For example, "acb dfe" is an anagram of "abc def", but "def cab" and "adc bef" are not.

Return the number of distinct anagrams of s. Since the answer may be very large, return it modulo 109 

+ 7..= ["a"]

Output: [["a"]]


Example 1:

Input: s = "too hot"
Output: 18
Explanation: Some of the anagrams of the given string are "too hot", "oot hot", "oto toh", "too toh", and "too oht".

Example 2:

Input: s = "aa"
Output: 1
Explanation: There is only one anagram possible for the given string.

Approach #1 - Using itertools permutations 

```python
import itertools

def generate_anagrams(s):
    # Generate all permutations of the string
    anagrams = set(itertools.permutations(s))
    # Convert tuples to strings and filter out those maintaining the space position
    anagrams = [''.join(p) for p in anagrams if p[3] == ' ']
    return anagrams

# Example usage:
anagrams = generate_anagrams("abc def")
print(anagrams[:10])  # Print first 10 anagrams
```

```python
def generate_anagrams_fixed_space(s, space_index):
    def backtrack(path, choices):
        if not choices:
            anagram = ''.join(path)
            if anagram[space_index] == ' ':
                anagrams.add(anagram)
        for i in range(len(choices)):
            backtrack(path + [choices[i]], choices[:i] + choices[i+1:])
    
    anagrams = set()
    backtrack([], list(s))
    return list(anagrams)

# Example usage:
anagrams = generate_anagrams_fixed_space("abc def", 3)
print(anagrams[:10])  # Print first 10 anagrams
```

```python
def generate_anagrams_backtracking(s):
    def backtrack(path, choices):
        if not choices:
            anagrams.add(''.join(path))
        for i in range(len(choices)):
            backtrack(path + [choices[i]], choices[:i] + choices[i+1:])

    anagrams = set()
    backtrack([], list(s))
    return list(anagrams)

# Example usage:
anagrams = generate_anagrams_backtracking("abc def")
print(anagrams[:10])  # Print first 10 anagrams
```

```python
def generate_anagrams(s, answer=""):
    if len(s) == 0:
        print(answer, end=" ")
        return
    
    for i in range(len(s)):
        char = s[i]
        # Removing char[i] from string
        ros = s[:i] + s[i+1:]
        # Recursive call
        generate_anagrams(ros, answer + char)

# Example usage:
generate_anagrams("abcdef")
```

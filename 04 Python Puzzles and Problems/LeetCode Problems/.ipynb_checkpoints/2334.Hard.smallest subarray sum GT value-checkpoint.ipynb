### Smallest Subarray with Sum greater than a given value 

Given an array of integers and a number x, the task is to find the smallest subarray with a sum strictly greater than X 

Example:
arr[1, 10, 5, 2, 7 ], X = 9
Minimum length subarray is {10}

Input : arr[] = {1, 4, 45, 6, 0, 19}, x  =  51
Output: 3
Minimum length subarray is {4, 45, 

Questions: 
1. Integer: +ve, -ve or Xero whole number
2. What if sum is LT given value
3. 6}

### Some boundary conditions to consider:

Minimum Length: The array could be as small as 1 element. Make sure to handle small arrays gracefully.

No Solution: It's possible that no subarray meets the criteria. Handle this case by returning an appropriate response, like -1 or an error message.

All Positive Numbers: The array can consist of all positive numbers. This simplifies the problem since you wonâ€™t need to deal with negative values reducing the sum.

Array with Negative Numbers: The array could contain negative numbers which might complicate the logic, especially if these numbers are between the start and end of the subarray.

Edge Values: The target sum might be close to the sum of the entire array, or it might be unattainably high.

Duplicates: The array might have duplicate elements which could affect the subarray sums in unpredictable ways.

### Naive Approach
    - use two nested loops
    -  The outer loop picks a starting element, the inner loop considers all elements (on right side of current start) as ending element. 
    - Whenever sum of elements between current start and end becomes more than the given number, update the result if current length is smaller than the smallest length so far. 

```python
#  Naive Approach 
def smallestSubarraySum(arr, n, target):
    curr_sum = 0
    subarray = []
    min_length = 0
    for start in range(0,n,1):         
        for end in range(start+1, n,1):
            # print (f"start={start}, arr[start]={arr[start]}, end={end}, curr_sum={curr_sum}, arr[end]={arr[end]}")            
            curr_sum += arr[end] 
            subarray.append(arr[end])
            min_length += 1
            if curr_sum >= target and min_length < n:
                return min_length, subarray            
    return -1
            

# arr = [1, 16, 45, 6, 0, 19]
# target = 51
# arr = [1, 6, 45]
# target = 7
arr = [1,3,4,3,1]
target = 6
n = len(arr)
print (smallestSubarraySum(arr, n, target))
# print (min_subarray)
```

```python
def smallest_subarray_with_sum(arr, target):
    n = len(arr)
    min_length = float('inf')
    current_sum = 0
    start = 0
    subarray = []

    for end in range(n):
        current_sum += arr[end]
        print (f"Iteration #{end}")
        print (f"current_sum={current_sum}, arr[end]={arr[end]}, target={target}")
        while current_sum > target:
            print (f"end - start + 1={end - start + 1}, min_length={min_length}")
            if end - start + 1 < min_length:
                min_length = end - start + 1
                subarray = arr[start:end + 1]
            current_sum -= arr[start]
            start += 1

    return (min_length, subarray) if min_length != float('inf') else (-1, [])

# Example usage:
# arr = [2, 3, 1, 2, 4, 3]
# target = 7

arr = [1, 16, 45, 6, 0, 19]
target = 51
min_length, subarray = smallest_subarray_with_sum(arr, target)

print(f"Minimum length: {min_length}")
print(f"Subarray: {subarray}")

# arr = [2, 3, 1, 2, 4, 3]
# target = 7
```

```python
def smallestSubArraywithSum(arr, n, x):
    # initialize he lengt of the smallest subarray as n+1
    min_len = n + 1
    # pick every element as starting point 
    for start in range(0, n):
        # initialize the sum starting with current start
        curr_sum = arr[start]
        # if the first elemet is grater 
        if (curr_sum > x):
            return 1
        # Now try different ending points for the current starting point 
        for end in range(start+1, n):
            # add last element to the current sum 
            curr_sum += arr[end]
            if (curr_sum > x) and (end - start + 1) < min_len:
                min_len = (end - start + 1)
    return min_len

arr1 = [1, 4, 45, 6, 10, 19]
x = 51
n1 = len(arr1)
res1 = smallestSubArraywithSum(arr1, n1, x);
if res1 == n1+1:
    print("Not possible")
else:
    print(res1)
```

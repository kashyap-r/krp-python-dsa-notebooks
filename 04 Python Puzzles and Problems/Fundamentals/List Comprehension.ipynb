```python
"""
List Comprehension
https://www.geeksforgeeks.org/python/python-list-comprehension/

https://www.geeksforgeeks.org/python/python-list-comprehension-using-if-else/

https://www.geeksforgeeks.org/python/python-list-comprehension/

Syntax of List Comprehension
[expression for item in iterable if condition]

expression: The transformation or value to be included in the new list.
item: The current element taken from the iterable.
iterable: A sequence or collection (e.g., list, tuple, set).
if condition (optional): A filtering condition that decides whether the current item should be included.
This syntax allows us to combine iteration, modification, and conditional filtering all in one line.
"""
```

```python
a = [2, 3, 4, 5]
res = [val ** 2 for val in a ]
print (res)
res2 = [val ** 3 for val in a]
print (res2)
```

```python
# Using a for loop
res = []
for val in a:
    res.append(val ** 2)
print (res)

print (list(val ** 2 for val in a))
"""
Explanation: In the above list comprehension, the iterable is a list 'a', and the expression is val * 2, which multiplies each value from the list by 2
"""
```

```python
"""
Conditional Statements in List Comprehension

"""
a = [1, 2, 3, 4, 5]
res = [val for val in a if val % 2 == 0]
print (res)
res = [val * 2 for val in a if val % 2 != 0]
print (res)
```

```python
# Creating a list from a range
a = [i for i in range(10)]
print (a)
b = list(i for i in range(10) if i % 2 == 0)
print (b)
```

```python
#Using the nested loops
# generate a list of coordinate pairs for a simple 3x3 grid
# Creates a list of tuples representing all combinations of (x, y)
# where both x and y range from 0 to 2.
coordinates = [(x, y) for x in range(3) for y in range(3)]
print(coordinates)
```

```python
# creating tables or matrices
# Using traditional for loop
matrix = []  # start with an empty list (our matrix)

for j in range(3):         # outer loop → rows
    row = []               # start a new row
    for i in range(1, 4):  # inner loop → columns
        value = i + j*3    # calculate value for this cell
        row.append(value)  # add it to the current row
    matrix.append(row)     # add the row to the matrix

print(matrix)

# now using list comprehensions

mat = [[]]
```

```python
# Flattening a list of lists i.e. converting a list of lists into a single list

# first generate a 3 x 3 matrix
matrix = [[i + j*3 for i in range(1, 4)] for j in range(3)]
print(matrix)
```

```python
# Using init function
"""
use the __inti__() function to initialize the object attributes when an object is created
the init() method is the constructor in python
"""
class Dog:
    species  = "Canine"  #class attribute

    # the IIinit__ method intiializes the name and age attributes when a new object is created
    def __init__(self, name, age):
        self.name = name # instance attribute
        self.age = age # instance attribute
    """
    the self parameter is a reference to the current instance of the class. It allows us to access the attributes and methods of the object
    """
    def bark(self):
        print(f"{self.name} is barking!")

# initiating object
dog1 = Dog("Buddy", 3)

print (dog1.name)
print (dog1.age)

"""
Inside bark(), self.name accesses the specific dog's name and prints it.
When we call dog1.bark(), Python automatically passes dog1 as self, allowing access to its attributes.
"""
dog1.bark()

```

```python
"""
THe ___str()___ method
__str__ Implementation: Defined as a method in the Dog class. Uses the self parameter to access the instance's attributes (name and age).
Readable Output: When print(dog1) is called, Python automatically uses the __str__ method to get a string representation of the object.

"""

class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __Str__(self):
        return f"{self.name} is {self.age} years old"
    def bark(self):
        print(f"{self.name} is barking!")

dog1 = Dog("Buddy", 3)
dog2 = Dog("Charlie", 12)
print(dog1.__str__())
print(dog2.__str__())
print (dog1.bark())
print (dog2.bark())
print (dog1)
print (dog2)
```

```python
"""
Class and Instance Variables

Class Variables
    -- are the variables shared across all the instances of a class
    -- defined at class level, outside any methods
    -- all objects of the class share the same value for a class variable

Instance Variables
    -- variables that are unique to each instance (object) of a class
    -- defined within the __init__() method or other instance methods
    - each object maintains its own copy of instance variables, independent of other objects
"""

class Dog:
    species = "Canine"

    def __init__(self, name, age):
        self.name = name
        self.age = age

dog1 = Dog("Charlie", 12)
dog2 = Dog("Zeus", 9)

# Access class and instance variables
print (dog1.species)
print (dog1.name)
print (dog1.age)

# Modify instance variables
dog1.name = "Max"
print (dog1.name)

# Modify class variables
Dog.species = "Feline"
print (dog1.species)
print (dog2.species)
cat1 = Dog("Zingy", 5)
print (cat1.species)
print (cat1.name)
# Changing the class variable impacts all the objects created under the class
print (dog2.species)
```

Class Variable (species): Shared by all instances of the class. Changing Dog.species affects all objects, as it's a property of the class itself.

Instance Variables (name, age): Defined in the __init__() method. Unique to each instance (e.g., dog1.name and dog2.name are different).

Accessing Variables: Class variables can be accessed via the class name (Dog.species) or an object (dog1.species). Instance variables are accessed via the object (dog1.name).

Updating Variables: Changing Dog.species affects all instances. Changing dog1.name only affects dog1 and does not impact dog2.


```python
# Getter and Setter methods

"""
- Provide controlled access to an objects attributes
- used to retrieve or modify the values of private attributes, allowing data encapsulation

Python does nto have explicity get and set methods, it supports this functionality using property decorators

Getter: Used to access the value of an attribute.
Setter: Used to modify the value of an attribute.

"""

class Dog:
    species = "Canine"
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @property
    def name(self):
        return self.name #gettr

    @name.setter
    def name(self, value):
        self.name = value

    @property
    def age(self):
        return self.age

    @age.setter
    def age(self, value):
        if value < 0:
            print ("Age cannot be negative")
        else:
            self.age = value

```

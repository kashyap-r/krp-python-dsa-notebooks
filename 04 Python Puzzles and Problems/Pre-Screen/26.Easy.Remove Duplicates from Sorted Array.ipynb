
26. Remove Duplicates from Sorted Array

https://leetcode.com/problems/remove-duplicates-from-sorted-array/description/

Problem Statement:

Given an integer array nums sorted in non-decreasing order, remove the duplicates in-place such that each unique element appears only once. The relative order of the elements should be kept the same. Then return the number of unique elements in nums.

Consider the number of unique elements of nums to be k, to get accepted, you need to do the following things:

    Change the array nums such that the first k elements of nums contain the unique elements in the order they were present in nums initially. The remaining elements of nums are not important as well as the size of nums.
    Return k.


Example: nums = [1, 1, 2]
OUtput: 2 

Edge Case Scenarios:
1. Input array is Empty 
2. Input array has unique elements 
3. Single Element array 
4. Array with all elements identical 


```python
# arr = [1, 1, 6]
arr = [1, 1, 2, 2, 3, 4, 5, 5, 6]
n = len(arr)
print (arr)
for i in range(n):      
    for j in range(i+1, n-1):
        if arr[i] == arr [j]:
            arr.pop(j)

print (arr)

```

```python
def remove_duplicates(nums):
    if not nums:
        return nums

    # Initialize the new index
    new_index = 1

    for i in range(1, len(nums)):
        if nums[i] != nums[i - 1]:
            nums[new_index] = nums[i]
            new_index += 1

    # Mark the remaining elements as None
    for i in range(new_index, len(nums)):
        nums[i] = None

    return nums

# Example usage:
nums = [1, 1, 2, 2, 3, 3, 4, 4]
remove_duplicates(nums)
print("Array after removing duplicates:", nums)  # Output: [1, 2, 3, 4, None, None, None, None]
```

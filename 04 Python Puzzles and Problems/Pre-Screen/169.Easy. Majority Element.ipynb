169.Easy. Majority Element

Given an array nums of size n, return the majority element.

The majority element is the element that appears more than ⌊n / 2⌋ times. You may assume that the majority element always exists in the array.

Example:
arr = [3, 2, 3]
Output: 3

Input: arr = [2, 2, 1, 1, 1, 2, 2]
Output: 2




Questions:
1. Is the array Sorted - Do we need to sort the array ?
2. Does the array contain onlu intergers or numeric values ?
3. What if there is a tie in the counts of majority elements 
4. All the elements of the array are identical 
5. Empty array 
6. array with one element

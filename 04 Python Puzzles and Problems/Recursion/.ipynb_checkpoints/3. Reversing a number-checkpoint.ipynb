```python
Reverse a number 
```

```python
# a pretty approach
def recRevNum(num, revNum=0):
    # base case: if num is a single digit number, return as-is 
    if num == 0:
        return revNum
    else: 
        revNum = revNum * 10 + num % 10 
        return recRevNum(num // 10, revNum)

def reverse_number(n):
    return recRevNum(n)
    
number = 1234
revNumber = reverse_number (number)
print (f"The number: {number}")
print (f"The reversed number:{revNumber}")
```

```python
# another not so pretty approach
def revNumber(num):
    if num < 10:
        return num 
    else:
        lastdigit = num % 10
        remNum = num // 10 
        revNum = revNumber(remNum)        
        numDigits = len(str(remNum))
        return lastdigit * (10 ** numDigits) + revNum
num = 1234 
print (f"number:{num}")
print (f"reversed number: {revNumber(num)}")
```

```python
div = 123
revNum = 0
while div != 0:
    rem = div % 10
    div = div // 10 
    print (f"remainder:{rem}")
    print (f"divisor:{div}")
    revNum = (revNum * 10) + rem
print (f"Now the number in reverse: {revNum}")
```

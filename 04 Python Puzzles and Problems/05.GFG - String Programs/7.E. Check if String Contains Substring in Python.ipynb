```python
# Using if-else

str = "May name is Kashyap"
if 'ash' in str:
    print ("True")
else:
    print ("False")
```

```python
# Using In operator
"""
Time Complexity : O(n)
Auxiliary Space : O(1)
"""
str = "The name is Kashyap"
flag = True if 'ash' in str else False
print (flag)
```

```python
# Checking using split() method
# this only checks for the entire word not for substrings inside the words in the list


str = "May name is Kashyap"

subs = "ash"

s = str.split()
print (s)
print (True if subs in s else False)

for word in s:
    print (True if subs in word else False)

"""
Time Complexity : O(n + m)
Auxiliary Space : O(n)
"""
```

```python
# Using find method
str = "May name is Kashyap"
if (str.find("ashu") == -1 ):
    print ("False")
else:
    print ("True")

"""
Time Complexity : O(N)
Auxiliary Space : O(1)
"""

```

```python
"""
Using count() method
"""
def check(s2, s1):
    if (s2.count(s1) > 0):
        print("YES")
    else:
        print("NO")


s2 = "A geek in need is a geek indeed"
s1 = "geek"
check(s2, s1)
print (s2.count(s1))

"""
Time Complexity : O(N)
Auxiliary Space : O(1)
"""
```

```python
"""
Using index() method
"""

str = "My name is Kashyap"

print (str.index('ash', 0, len(str)))


"""
Time Complexity : O(N)
Auxiliary Space : O(1)
"""
```

```python
# Using List comprehension
str = "My name is Kashyap"
sub = "ash"
print (["Yes" if sub in str else "no"])

"""
Time Complexity : O(N)
Auxiliary Space : O(1)
"""
```

```python
# using Lambda Function

s = "My name is Kashyap"
s2 = "ash"
#
# x = list(filter(lambda x: (subs in str, str.split())))
# print(['yes' if x else "no"])
#
# s="geeks for geeks"
# s2="geeks"
x=list(filter(lambda x: (s2 in s),s.split()))
print(["yes" if x else "no"])

"""
Time Complexity : O(n + m)
Auxiliary Space : O(m)
"""
```

```python
"""
Python - Ways to Count Number of Substring in String

Given a string S, determine the number of substring that satisfy a certain criteria. For example, given S = "abcdeabcpqabcxuz", to find the substring = abc the function should return 3.
"""
```

```python
#Using str.count()

str = "abcdeabcpqabcxuz"
fstr = "abc"

cnt = str.count('abc')
print (cnt)
```

```python
"""
Using for loop with string slicing - kind of sliding window
O(t) = n
O(S) = 1
"""

str = "abcdeabcpqabcxuz"
fstr = "abc"
cnt = 0

for i in range (len(str) - len(fstr) + 1):
    if str[i:i+len(fstr)] == fstr:
        cnt += 1

print (cnt)
```

```python
"""
Using list comprehension
"""

# Input string and substring to count
s = "hello world, hello universe"
str = "abcdeabcpqabcxuz"
sub = "hello"

# Count occurrences of the substring using a generator expression
c = sum(1 for i in range(len(s) - len(sub) + 1) if s[i:i + len(sub)] == sub)

print(c)
```

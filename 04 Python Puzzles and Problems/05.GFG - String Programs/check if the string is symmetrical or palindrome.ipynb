Problem Statement

Given a string, check if the string is symmetrical and palindrome 

Def: A string is "Symmetrical" if both halves of the string are the same 
Ex: KhoKho 
Def: A string is "palindrome" if one hald of the string is the reverse of the other half.
Ex: Malayalam 

Edge Cases: 
1. Empty string 
2. String length <= 2 
3. Input string has special characters 
4. String has numbers 


Approach 
1. Naive Approach 
TO Check for palindrome, we can use the loop to compare characters from the start and end moving towards the middle 
For checking SYmmetry, we split the string into two halves and compare them 


```python
def palindrome(s):
    return s == s[::-1]

def isSymmetrical(s):
    n = len(s)
    mid = n // 2 
    if n % 2 == 0:
        return s[:mid] == s[mid:]
    else:
        return s[:mid] == s[mid+1:]
    
s = 'ohkkho'
s = 'khokho'

if isSymmetrical(s):
    print ("Is symmetrical")
else:
    print ("Is not symmetrical")

if palindrome(s):
    print ("Is palindrome")
else:
    print ("Not a palindrome")


"""
Time COmplexity O(n)
Space Complexity O(n)
"""
```

```python
# Approach #2- Using string slicing 

s = "amaama"
mid = len(s) // 2
s1 = s[:mid]
s2 = s[mid:] if len(s) % 2 == 0 else s[mid+1:]

if s1 == s2:
    print ("Symmetrical")
else:
    print ("not symmetrical")
```

```python
# Approach #3 - Using regular expression 
import re

def palindrome(s):
    s1 = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    return s1 == s1[::-1]

def symmetrical(s):
    s2 = re.sub(r'\s+', '', s)
    n = len(s2)
    mid = n // 2 
    if mid % 2 == 0:
        return s2[:mid] == s2[mid:]
    else:
        return s2[:mid] == s2[mid+1:]
    

s = "madam"
print(f"Is '{s}' Palindrome? {palindrome(s)}")
print(f"Is '{s}' Symmetrical? {symmetrical(s)}")

s = "abba"
print(f"Is '{s}' Palindrome? {palindrome(s)}")
print(f"Is '{s}' Symmetrical? {symmetrical(s)}")
```

```python
# Using reversed() Function

def isPalindromeRevresed(s):
    s1 = ''.join(reversed(s))
    return s == s1

def isSymmetrical(s):
    mid = len(s) // 2
    if len(s) % 2 == 0:
        s1 = s[:mid]
        s2 = s[mid:]
    else:
        s1 = s[:mid]
        s2 = s[mid + 1:]
    return s1 == s2 

str = ["khokho", "amaama", "malayalam"] 

for s in str:
    print(f"Testing string: {s}")
    
    if isSymmetrical(s):
        print("Symmetrical")
    else:
        print("Not a Symmetrical")
    
    if isPalindromeRevresed(s):
        print("Palindrome")
    else:
        print("Not a Palindrome")


"""
Time Complexity: O(n)
Auxiliary Space: O(n)
"""
```

### Slicing with Negative numbers

```python
# Slicing 

text = "GeeksforGeeks"

# Slicing left part of string text
left = text[:-8]

# Slicing middle part of string text
middle = text[-8:-5]

# Slicing right part of string text
right = text[-5:]

print(left)
print(middle)
print(right)
```

```python
# slice function 

text = 'GeeksforGeeks'

# Slicing left part of string text
left = text[slice(-8)]

# Slicing middle part of string text
middle = text[slice(-8,-5)]

# Slicing right part of string text
right = text[slice(-5,None)]

print(left)
print(middle)
print(right)
```

<!-- String join() Method -->

The join() method us ised to concatenate the elements of an iterable (list, tuple or set) into a single string with a specified delimiter placed between each element. 

```python

arr = ["Hello", "world", "from", "python"]

res = ' '.join(arr)
print (res)
type(res)
```

```python
# join with tuples 

s = ("I", "am", "learning", "to", "code")
print ('-'.join(s))
```

```python
# join with set 
s = {'Python', 'is', 'fun'}
print ('-'.join(s))
```

```python
# Using join() with Dictionary 
# 

d = {'I am': 1, 'a Geek': 2, 'Heehaw': 3}

print (d)
print (d.keys())
print (d.values())

print ('-'.join(d))
```

```python
# using map function
s = "geeks,for,geeks"


# Split and convert each item to uppercase using map
res = list(map(str.upper, s.split(',')))

# Print each item
for item in res:
    print(item)
```

```python
s = "geeks,for,geeks"

# Split at the first comma using partition
a, _, b = s.partition(',')

print(a)  # Part before the first comma ('geeks')
print(b)  # Part after the first comma ('for,geeks')
```


## Program to sort an array of strings using Selection Sort

Reference:
https://www.geeksforgeeks.org/program-to-sort-an-array-of-strings-using-selection-sort/?ref=next_article

Problem Statement: Given an array of strings, sort the array using Selection Sort.

Examples: 
Input  : paper true soap floppy flower
Output : floppy, flower, paper, soap, true

Considerations / Edge Cases: 
1. Case sensitive ?
2. what if strings are same and of different case 
3. Contains alphanumeric characters
4. Contains special characters 
5. Contains escape spaces, sequences 
6. Input string is empty or just has one string
7. Input string contains numbers, decimals
8. Input string Size limits 


```python
def stringSelectionSort(arr, n):
    for i in range(n):
        min_index = i 
        min_str = arr[i]
        # find the min element in the unsorted array
        for j in range(i+1, n):
            if min_str > arr[j]:
                min_str = arr[j]
                min_index = j 
        # swap the found min element with the first element
        if min_index != i:
            temp = arr[i]
            arr[i] = arr[min_index]
            arr[min_index] = temp 
        return arr 

arr = ["GeeksforGeeks", "Practice.GeeksforGeeks", "GeeksQuiz"]
 
print("Given array is")
for i in range(len(arr)):
    print(i,":",arr[i])
 
print("\nSorted array is")
for i in range(len(stringSelectionSort(arr,len(arr)))):
    print(i,":",stringSelectionSort(arr,len(arr))[i])
```

Time Complexity: O(n2), where n represents the size of the character array.
Auxiliary Space: O(100), no extra space is required, so it is a constant.

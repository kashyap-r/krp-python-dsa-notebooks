```python
"""
Strings in Python

A string is a sequence of characters. Python treats anything inside quotes as a string. This includes letters, numbers, and symbols. Python has no character data type so single character is a string of length 1.
"""
```

```python
s = "Kashyap"
print (s[1])
s1 = s + s[1]
print (s1)
```

```python
# Strings can be created using either single or double quotes
s1 = 'kash'
s2 = "kash"
print (s1)
print (s2)
# Multi-line string
s3 = """My Name is Kashyap. I am from Bengaluru. I am a data engineer and I love travelling"""
print (s3)
# Multi-line string can also be enclosed in singe quotes
s4 = '''My Name is Kashyap. I am from Bengaluru. I am a data engineer and I love travelling'''
print (s4)
```

```python
"""
Accessing characters in string
"""

str = "Kashyap"

print (str[0])
print (str[-1])
str[-4]
```

```python
"""
String SLicing
Slicing is a way to extract portion of a string by specifying the start and end indexes. The syntax for slicing is string[start:end], where start starting index and end is stopping index (excluded).

Note: refer the relevant notebook for string slicing and dicing
"""
```

```python
"""
String Immutability
- Strings are immutable
- this means they cannot be changed once they're created
"""

str = "Kashyap"
# str[0] = "S"
print (str)
#TypeError: 'str' object does not support item assignment

s = 'B' + str[1:]
print (s)

# deleting a string
# del s
print (s)
# NameError: name 's' is not defined

# Updating the string
s = "Hello, I'm here!!"
print (s)
s1 = s + ", Where are you?"
print (s1)

s2 = s.replace('Hello', 'Kashyap')
print (s2)
```

```python
"""
String Methods
upper() and lower()
"""
s = "Kashyap, I'm here"
print (s)
print (s.upper())
print (s.lower())

"""
strip() abd replace()
strip() - removes leading and trailing whitespace from the string
replace() - replaces all occurances of a specified substring with another
"""

s = "   Kashyap, I'm here!"
s1= "Kashyap, I'm here!    "
print (s.strip())
print (s1.strip())

print (s.replace(" ", "X"))
print (s.replace(' ', "Z"))
```

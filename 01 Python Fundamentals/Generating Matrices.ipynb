```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(matrix)
```

```python
matrix = []  # start with an empty list (our matrix)

for j in range(3):         # outer loop → rows
    row = []               # start a new row
    for i in range(1, 4):  # inner loop → columns
        value = i + j*3    # calculate value for this cell
        row.append(value)  # add it to the current row
    matrix.append(row)     # add the row to the matrix

print(matrix)
```

```python
mat = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

res = [val for row in mat for val in row]

print(res)
```

```python
# Using list comprehension

matrix = [[i+j*3 for i in range(1,4)] for j in range(3)]
print(matrix)
```

```python
# Matrix of random number
import random
matrix = [[random.randint(1, 100) for _ in range(3)] for _ in range(3)]
print(matrix)
```

```python
# With Numpy
import numpy as np

matrix = np.arange(1, 10).reshape(3, 3)
print(matrix)
```

```python
# Multiplication table (3×3)

matrix = []
for j in range(1, 4):        # rows
    row = []
    for i in range(1, 4):    # columns
        value = i * j        # multiplication table logic
        row.append(value)
    matrix.append(row)

print(matrix)
```

```python
import random

matrix = []
for j in range(3):
    row = []
    for i in range(3):
        value = random.randint(1, 100)
        row.append(value)
    matrix.append(row)

print(matrix)
```

```python
# Letters (A to I

matrix = []
ascii_A = ord('A')  # 65
for j in range(3):
    row = []
    for i in range(3):
        value = chr(ascii_A + j*3 + i)  # convert number to letter
        row.append(value)
    matrix.append(row)

print(matrix)
```

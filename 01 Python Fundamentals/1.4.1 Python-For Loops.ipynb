## For Loops in Python.

How to use the for loop in Python?

    In Python, the for loop is used to iterate over a sequence (such as a list, tuple, string, or dictionary) or any iterable object

Syntax: 

for var in iterable:
    # statements
    pass

### Python For Loop with String

This code uses a for loop to iterate over a string and print each character on a new line. 

```python
# Iterating over a String
print("String Iteration")

s = "Go Kashyap-Go"
for i in s:
    print(i)
```

### Python for loop with Range

This code uses a Python for loop with index in conjunction with the range() function to generate a sequence of numbers starting from 0, up to (but not including) 10, and with a step size of 2

```python
for i in range(0, 10, 2):
    print(i)
```

```python
for i in range(5):
    print(i)
```

### Python for loop Enumerate

In Python, the enumerate() function is used with the for loop to iterate over an iterable while also keeping track of the index of each item.

```python
list1 = ['K', 'a', 's', 'h', 'y', 'a', 'p']

for count, ele in enumerate(list1):
    print (count, ele)
```

### Nested For Loops in Python

<!-- Say to print a square matrix of the order 3 -->

```python
for i in range(3):
    for j in range(3):
        print ('1')
    print()
```

### Python For Loop Over List

use a for loop to iterate over a list of strings 

```python
list2 = ["Kashyap", "Raj", "Purohit"]
for str in list2:
    print (str)
```

### Python for loop in One Line

```python
numbers = [x for x in range(10)]
print (numbers)
```

### Python For Loop with Dictionary


```python
d = dict()
d[1] = 'Kash'
d[2] = "Raj"
print (d)

for i in d:
    print (f"i = {i}, d[i] = {d[i]}")
    
```

```python
# Iterating over dictionary
print("Dictionary Iteration")

d = dict()

d['xyz'] = 123
d['abc'] = 345
for i in d:
    print("% s % d" % (i, d[i]))
```

### Python For Loop with Tuple


```python
tup = ((1, 2), (3, 4), (5, 6))
for t in tup:
    print (t)
```

```python
t = ((1, 2), (3, 4), (5, 6))
for a, b in t:
    print(a, b)
```

### Python For Loop with Zip()

This code uses the zip() function to iterate over two lists (fruits and colors) in parallel. The for loop assigns the corresponding elements of both lists to the variables fruit and color in each iteration.

```python
fruits = ["apple", "banana", "cherry"]
colors = ["red", "yellow", "green"]
for fruit, color in zip(fruits, colors):
    print(fruit, "is", color)
```

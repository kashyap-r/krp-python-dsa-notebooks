```python
#Return Calculator 
def rcalc(a, b):
    return (a + b)
print ('The sum is ',rcalc(2,3))
```

```python
#Full Names 

def fname(fname, lname):
    print ('My name is', fname)
    print ('My full name is ..')
    print (fname + ' ' + lname)

fname ('James','Bond')
fname ('Peter','Parker')
fname ('Bruce','Wayne')
```

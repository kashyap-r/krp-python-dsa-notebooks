# Dictionary in Python 
- it consists of key-value pairs
- in python dictionary is similar to hash or maps in other languages

```python
d = dict()
d['xyz'] = 123
d['abc'] = 345
print (d)
print (f"The dictionary keys: {d.keys()}")
print (f"THe dictionary values: {d.values()}")
for i in d:
    print ("%s %d" %(i, d[i]))

for index, key in enumerate(d):
    print (index, key, d[key])

print ('abc' in d)
del d['abc']
print ('abc' in d)
```

Time Complexity: O(1)
Auxiliary Space: O(n)

### break, continue, pass in Python 

* break - takes you out of current loop
* continue - ends the current iteration in the loop and moves to the next iteration
* pass - does nothing

```python
# Function to illustrate break in loop 
def breakTest(arr): 
    for i in arr: 
        if i == 5: 
            break
        print (i) 
    # For new line 
    print("")  
  
# Function to illustrate continue in loop 
def continueTest(arr): 
    for i in arr: 
        if i == 5: 
            continue
        print (i) 
  
    # For new line 
    print("")  
  
# Function to illustrate pass  
def passTest(arr): 
        pass
      
# Driver program to test above functions 
# Array to be used for above functions: 
arr = [1, 3 , 4, 5, 6 , 7] 
  
# Illustrate break 
print ("Break method output") 
breakTest(arr) 
  
# Illustrate continue  
print ("Continue method output") 
continueTest(arr) 
  
# Illustrate pass- Does nothing 
passTest(arr)
```

Time complexity: O(n), where n is the length of the array arr.
Auxiliary space: O(1), since the space used is constant and does not depend on the size of the input.

### map, filter, lambda

**map:** the map() fuction applies a function to every memeber of the iterable and returns the result.
If there are multiple arguements, map() returns a list consisting of tuples containing the corresponding items from all iterables. 

**filter:** It takes a function returning True or False and applies it to a sequence, returning oa list of only those members of the sequemce for which the function returned True. 

**lambda:** Python provides the ability to create a simple anaonymous inline function called lambda function. Using lambda and map you can have two for loops in one line. 


```python
items = [1, 2, 3, 4, 5]

cubes = list(map(lambda x: x**3, items))
print (cubes)
print( (lambda x: x**2)(5)) 
for i in items:
    print( (lambda x: x**2)(i), end=' ') 
print()
# Make function of two arguments that return their product 
print ((lambda x, y: x*y)(3, 4))

```

```python
#Using filter function to filter all  
# numbers less than 5 from a list 
number_list = range(-10, 10) 
print (number_list)
less_than_five = list(filter(lambda x: x < 5, number_list)) 
print(less_than_five)
```

```python
# code without using map, filter and lambda 
  
# Find the number which are odd in the list 
# and multiply them by 5 and create a new list 
  
# Declare a new list 
x = [2, 3, 4, 5, 6] 
  
# Empty list for answer 
y = [] 
  
# Perform the operations and print the answer 
for v in x: 
    if v % 2: 
        y += [v*5] 
print(y) 

# above code with map, filter and lambda 
  
# Declare a list 
x = [2, 3, 4, 5, 6] 
  
# Perform the same operation as  in above post 
y = list(map(lambda v: v * 5, filter(lambda u: u % 2, x))) 
print(y) 
```

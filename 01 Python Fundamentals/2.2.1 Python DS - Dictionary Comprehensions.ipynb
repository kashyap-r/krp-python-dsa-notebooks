### Python Dictionary Comprehension

- Allows keys and values to be generated from a loop in one line

```python
sq = {x: x**2 for x in range(5)}
print(sq)
```

### Creating a dictionry from two lists

```python
# using zip()

keys = ['a', 'b', 'c', 'd']
values = ['one', 'two', 'three', 'four']

d = {k:v for (k, v) in zip(keys, values)}
print(d)
```

```python
# using fromkeys() method

d = d.fromkeys(range(5), True)
print (d)
```

```python
b = {x.upper(): x*3 for x in 'coding'}
print (b)
```

```python
b = {x.upper(): x for x in 'coding'}
print (b)
```

### Dictionary Comprehension with conditional Statements

We can include conditions in a dictionary comprehension to filter items or apply logic only to specific values
This allows us to create dictionaries selectively

```python
d = {x: x**3 for x in range (10) if x**3 % 4 == 0}
print (d)
```

### Nested Dictionary Comprehension

Creating nested dictionaries using dictionary comprehension


```python
s = 'KR'

res = {x: {y:x + y for y in s} for x in s}

print (res)
```

# Find all distinct elements in a given array 

input array arr = [2, 2, 3, 3, 7, 5]
Output: [2, 3, 7, 5]

input arr = [1, 2, 3, 4, 5]
output: [1, 2, 3, 4, 5]


```python
# Naive approach using nested loops 
"""
O(n^2) time and O(1) space
"""


def remDuplicate(arr):
    res = []
    for i in range(len(arr)):
        j = 0 
        while j < i:
            if arr[i] == arr[j]:
                break 
            j += 1
        
        if i == j:
            res.append(arr[i])

    return res


if __name__ == "__main__":
    arr = [2, 2, 3, 3, 7, 5]
    res = remDuplicate(arr)

    for val in res:
        print (val, end=' ')
```

```python
# approach #2 - Using Sorting - O(n*log n) Time and O(1) space 
"""
"""
def remDuplicate(arr):
    res = []
    n = len(arr)

    arr.sort()

    for i in range(n):
        if i == 0 or arr[i] != arr[i-1]:
            res.append(arr[i])

    return res 

if __name__ == "__main__":
    arr = [2, 2, 3, 3, 7, 5]
    res = remDuplicate(arr)
    for ele in res:
        print (ele, end=" ")
```

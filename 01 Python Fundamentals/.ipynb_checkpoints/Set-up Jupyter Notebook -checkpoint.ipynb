```python
# Enable the toolbar: 
!jt -t [theme name] -T

# Enable the filename and logo: 
!jt -t [theme name] -N

# Enable the kernel logo: 
!jt -t [theme name] -kl

# To change the theme and enable the toolbar, filename, logo, kernel logo in same command line statement: 
!jt -t [theme name] -T -N -kl
```
